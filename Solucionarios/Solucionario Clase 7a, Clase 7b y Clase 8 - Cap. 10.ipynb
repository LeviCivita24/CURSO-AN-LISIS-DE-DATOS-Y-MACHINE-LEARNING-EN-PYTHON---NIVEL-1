```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x = np.random.uniform(0,2,100)
y = np.random.uniform(0,2,100)
x1 = x[y>1]; y1 = y[y>1]
x2 = x[y<1]; y2 = y[y<1]
```

```python
#Gráficos
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.scatter(x1,y1, label='Mayores a 1')
ax.scatter(x2,y2, label='Menores a 1')
ax.legend(loc='upper center')
ax.set_xticks([0,1/2,1,3/2,2])
ax.set_yticks([0,1/2,1,3/2,2])
fig.show()
```

## <font color='green'>**Ejercicio 16** </font>

Agregue una leyenda, cambie los colores a azul, rojo y negro, y modifique el estilo de línea a gusto, para el ejercicio resuelto 10.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## <font color='green'>**Ejercicio 17** </font>

Para el siguiente conjunto de datos, genere una gráfica de barras:

&emsp;&emsp;&emsp; $hombres = [20,35,30,35,27]$   
&emsp;&emsp;&emsp; $mujeres = [25,32,34,20,25]$

```python
import matplotlib.pyplot as plt
```

## <font color='green'>**Ejercicio 18** </font>

Cambie el entorno gráfico a $bmh$ para el ejercicio propuesto 12.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('bmh')
```

```python
#Datos
x = np.linspace(0,1,100)
y1 = x
y2 = x**2
y3 = np.sqrt(x)
```

```python
#Gráficos
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y1, color='b', linestyle='-', label='$x$')
ax.plot(x,y2, color='r', linestyle='--', label='$x^2$')
ax.plot(x,y3, color='k', linestyle='-.', label='$\sqrt{x}$')
ax.set_xlabel('Datos de Entrada')
ax.set_ylabel('Datos de Salida')
ax.legend(facecolor='white')
fig.show()
```

## <font color='green'>**Ejercicio 19** </font>

Dibuje las curvas del ejercicio resuelto 6 en una gráfica de tres ejes. No olvide los respectivos títulos y las etiquetas de cada eje.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x = np.linspace(0,1,100)
y1 = x
y2 = x**2
y3 = np.sqrt(x)
```

```python
#Gráficos
fig, ax = plt.subplots(1,3, figsize=(12,4))
fig.subplots_adjust(wspace=0.25)

ax[0].plot(x,y1, color='b', linestyle='-')
ax[0].set_xlabel('$x$')
ax[0].set_ylabel('$y$')
ax[0].set_title('$y_1=x$')

ax[1].plot(x,y2, color='r', linestyle='--')
ax[1].set_xlabel('$x$')
ax[1].set_ylabel('$y$')
ax[1].set_title('$y_2=x^2$')

ax[2].plot(x,y3, color='k', linestyle='-.')
ax[2].set_xlabel('$x$')
ax[2].set_ylabel('$y$')
ax[2].set_title('$y_3=\sqrt{x}$')

fig.show()
```

## <font color='green'>**Ejercicio 21** </font>

Visualmente, ¿cuál es la diferencia entre una distribución uniforme y una normal?. Genere una figura con dos ejes para ello.

***Ayuda 1:*** Consulte el método `random.uniform`.   
***Ayuda 2:*** Consulte el método `random.normal`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x1 = np.random.uniform(-1,1,100000)
x2 = np.random.normal(0,1,100000)
```

```python
fig, ax = plt.subplots(1,2, figsize=(8,4))
ax[0].hist(x1, bins=100)
ax[0].set_title('Uniforme')
ax[1].hist(x2, bins=100)
ax[1].set_title('Normal')
fig.show()
```
