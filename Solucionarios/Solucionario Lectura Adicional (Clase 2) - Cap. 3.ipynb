## <font color='green'>**Ejercicio 14** </font>

Con la siguiente lista de números, que representan las notas de dos
estudiantes a lo largo del año:

&emsp;&emsp;&emsp; $notas = [[3.5, 4.0, 3.8, 4.5], [5.0, 4.8, 4.5, 4.0], [2.0,3.8,4.5,1.0]]$

1. Calcule la nota promedio de cada estudiante.   
2. Calcule la nota promedio del grupo

```python
notas = [[3.5,4.0,3.8,4.5], [5.0,4.8,4.5,4.0], [2.0,3.8,4.5,1.0]]

est_1 = notas[0]
est_2 = notas[1]
est_3 = notas[2]

prom_1 = sum(est_1)/len(est_1)
prom_2 = sum(est_2)/len(est_2)
prom_3 = sum(est_3)/len(est_3)
prom = (prom_1+prom_2+prom_3)/3

print(f'Promedio Estudiante 1: {prom_1:0.1f}')
print(f'Promedio Estudiante 2: {prom_2:0.1f}')
print(f'Promedio Estudiante 3: {prom_3:0.1f}')
print(f'Promedio Grupo: {prom:0.1f}')
```

## <font color='green'>**Ejercicio 15** </font>

Escriba un programa que encuentre cuales son los elementos no comunes entre dos listas.

```python
lista_1 = [1,2,3,4,5,6,7]
lista_2 = [2,4,6,8,10]

interseccion = set(lista_1) & set(lista_2)
union = set(lista_1) | set(lista_2)
list(union - interseccion)
```
