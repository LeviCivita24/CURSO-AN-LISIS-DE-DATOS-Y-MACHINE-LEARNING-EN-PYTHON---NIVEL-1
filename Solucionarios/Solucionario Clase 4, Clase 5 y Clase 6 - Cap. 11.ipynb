## <font color='green'>**Ejercicio 10** </font>

Cargue el archivo *Trabajadores.xlsx* y efectúe:

1. Cree un dataframe con esta base de datos.
2. Quíteles las tildes a las palabras de la columna $Nombre$ y deje únicamente las primeras letras en mayúscula.
3. Ponga la primera letra de las palabras de la columna $Concepto$ en mayúscula.
4. Separe la tabla en diferentes dataframes, uno por cada nombre.
5. Guarde los dataframes en un achivo de Excel y por cada nombre cree una hoja, incluya la base de datos original.

***Ayuda 1:*** Puede que le sean útiles los métodos $str.title$ y $apply$.     
***Ayuda 2:*** Use los resultados de los ejercicios resueltos 8.6 y 11.2.      
***Ayuda 3:*** Puede que le sea útil el método $groupby$.
***Ayuda 3:*** Puede que le sea útil el método $get\_group$.   
***Ayuda 5:*** Puede que le sea útil el método $ExcelWriter$.    
***Ayuda 6:*** Puede que le sea útil el método $to\_excel$.    

```python
import pandas as pd
```

```python
#Pregunta 1
df = pd.read_excel('Trabajadores.xlsx', sheet_name='Ventas')
```

```python
#Pregunta 2 y 3
def quitar_tildes(palabra):
    vocales = [['á','a'],['é','e'],['í','i'],['ó','o'],['ú','u']]
    for a,b in vocales:
        palabra = palabra.replace(a,b)
    return palabra

df.loc[:, 'Nombre'] = df.loc[:, 'Nombre'].apply(quitar_tildes)
df.loc[:, 'Nombre'] = df.loc[:, 'Nombre'].apply(str.title)
df.loc[:, 'Concepto']= df.loc[:, 'Concepto'].apply(str.title)
df
```

```python
#Pregunta 4 y 5
nombres = set(df.loc[:, 'Nombre'])
archivo = pd.ExcelWriter('Ventas.xlsx')
df.to_excel(archivo, sheet_name='Ventas', index=False)
grupos = df.groupby('Nombre')

for nombre in nombres:
  df1 = grupos.get_group(nombre)
  df1.to_excel(archivo, sheet_name=nombre, index=False)

archivo.close()
```

```python
#Quitar Tildes
def quitar_tildes(palabra):
    vocales = [['á','a'],['é','e'],['í','i'],['ó','o'],['ú','u']]
    for a,b in vocales:
        palabra = palabra.replace(a,b)
    return palabra
```

```python
#Pregunta 7
df = df.reindex(columns=['Nombre','Correo','Puntaje','Nota','Aprobó'])
df.head(5)
```

## <font color='green'>**Ejercicio 21** </font>

Escriba una función para traducir a español, una lista con palabras.

***Ayuda:*** Use la librería googletrans. Consulte.

```python
!pip install googletrans==3.1.0a0
```

```python
from googletrans import Translator
```

```python
def traducir(lista):
  lista_traducida = []

  for palabra in lista:
    traductor = Translator()
    t = traductor.translate(palabra, src='en', dest='es')
    lista_traducida.append(t.text)

  return lista_traducida

traducir(['One', 'Two', 'Three', 'Fourth', 'Five'])
```
