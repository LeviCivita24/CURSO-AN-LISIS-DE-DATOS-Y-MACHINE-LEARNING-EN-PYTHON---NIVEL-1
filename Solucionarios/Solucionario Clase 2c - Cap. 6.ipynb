## <font color='green'>**Ejercicio 13** </font>

Calcule, usando ciclos for, los primeros ocho números primos.

***Ayuda:*** Primeros ocho números primos: $2$, $3$, $5$, $7$, $11$, $13$, $17$, $19$.

```python
n = 20
primos = []

for numero in range(2,n+1):
  contador = 0

  for i in range(2,numero):
    if numero%i== 0:
      contador += 1

  if contador==0:
    primos.append(numero)

primos
```

## <font color='green'>**Ejercicio 17** </font>

Ponga en marcha el ejercicio propuesto 3.14 usando ciclos for.

```python
notas = [[3.5,4.0,3.8,4.5], [5.0,4.8,4.5,4.0], [2.0,3.8,4.5,1.0]]
prom = 0

for i in range(3):
  est_i = notas[i]
  prom_i = sum(est_i)/len(est_i)
  print(f'Promedio Estudiante {i}: {prom_i:0.1f}')

  prom += prom_i

prom /= 3
print(f'Promedio Grupo: {prom:0.1f}')
```

## <font color='green'>**Ejercicio 19** </font>

Lleve a cabo un descuento a un conjunto de precios siguiendo las reglas del ejercicio resuelto 5.6.

```python
precios = [1000,2500,5000,7500,10000,12500]
precios_d = []

for p in precios:
  if p>10000:
    r = p-p*(20/100)
  elif p>=5000 and p<=10000:
    r = p-p*(10/100)
  else:
    r = p

  precios_d.append(r)

print(f'Descuentos: {precios_d}')
```
