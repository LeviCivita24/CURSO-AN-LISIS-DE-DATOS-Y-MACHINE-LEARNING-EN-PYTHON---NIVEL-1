## <font color='green'>**Ejercicio 30** </font>

Verificar si la expresión $\sqrt{a+b}=\sqrt{a}+\sqrt{b}$, es correcta. Escribir un código que lleve a cabo esta tarea.


```python
a = float(input('Ingrese a: '))
b = float(input('Ingrese b: '))
izq = (a+b)**0.5
der = a**0.5 + b**0.5
cond = izq==der

print(f'¿(a+b)^1/=a^1/2+b^1/2?: {cond}')
print(f'{izq:.1f}={der:.1f}')
```
