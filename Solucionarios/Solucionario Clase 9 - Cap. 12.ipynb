Cargue la base de datos $titanic$ y efectúe:

1. Genere un histograma con la columna $age$.
2. Use el parámetro $hue$ para categorizar por $sex$.
3. Use el parámetro $stat$ con estado $percent$. ¿Qué sucede?
4. Use el parámetro $multiple$ con estado $layer$. ¿Qué sucede?
5. Cambie todas las etiquetas al español.
6. ¿Qué puede concluir con este diagrama?

***Ayuda 1:*** Puede que le sea útil el graficador $histplot$.   
***Ayuda 2:*** Implemente los métodos de personalización de Matplotlib.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Datos
df = sns.load_dataset('titanic')
```

```python
#Gráficos
fig, ax = plt.subplots(1,1, figsize=(4,4))

ax = sns.histplot(data=df, x='age', bins=16, hue='sex', stat='percent', multiple='layer')
ax.set_xlabel('Edad')
ax.set_ylabel('Porcentaje')
ax.set_yticks([0,3,6,9,12])
ax.legend(['Mujer','Hombre'], title='Sexo')

fig.show()
```

```python
#Guardar
fig.savefig('Ejercicio_R2.pdf', bbox_inches='tight', pad_inches=0.2)
```

Del diagrama se puede concluir que:

1. El promedio de edad, tanto para hombres como para mujeres, se encontraba en un rango entre los $20$ y los $40$ años.
2. Los hombres fueron la categoría mayoritaria, excepto para un rango de edades entre $10$ y $15$ años.
3. Ambos sexos tenían distribuciones de edad similares, siendo la de los hombres más alta por su mayor cantidad.
4. Los mayores de $65$ años representaron la fracción minoritaria del total de personas presentes en el barco.
5. Los niños o infantes, entre $0$ y $5$ años, superanron en cantidad a los adultos mayores.

¿Qué diferencia una variable numérica de una categórica?

1. **Numérica:** Una variable numérica es aquella que permite caracterizar una serie de datos por medio de números.   

  Ejemplo: Estaturas de un grupo de personas.

2. **Categótica:** Una variable categórica es aquella que permite clasificar una serie de datos por medio de cualidades.

  Ejemplo: Separar un grupo de personas por sus estaturas.

## <font color='green'>**Ejercicio 5** </font>

Con la base de datos del ejercicio resuelto 12.2, efectúe:

2. Use el parámetro $hue$ para categorizar por $sex$.
3. Use el parámetro $style$ con estado $alive$. ¿Qué sucede?
4. Use el parámetro $size$ con estado $class$. ¿Qué sucede?
5. Cambie todas las etiquetas al español.

***Ayuda:*** Puede que le sea útil el graficador $scatterplot$.   

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Datos
df = sns.load_dataset('titanic')
df = df.loc[:300, :]
```

```python
#Gráficos
fig, ax = plt.subplots(1,1, figsize=(4,4))

ax = sns.scatterplot(data=df, x='age', y='fare', hue='sex', style='alive', size='class')
ax.set_xlabel('Edad')
ax.set_ylabel('Precio del Pasaje')
ax.set_xticks([0,20,40,60,80])

marcadores, etiquetas = ax.get_legend_handles_labels()
etiquetas_2 = ['Sexo','Hombre','Mujer','Clase','1ra','2da','3ra','Sobrevivió','No','Sí']
ax.legend(marcadores, etiquetas_2, loc='center', bbox_to_anchor=(1.25,0.50))

fig.show()
```

```python
#Guardar
fig.savefig('Ejercicio_R5.pdf', bbox_inches='tight', pad_inches=0.2)
```

Del diagrama se puede concluir que:

1. De forma general sin importar la edad, la clase o el sexo, el precio del boleto está por debajo de los $100$ dolares por persona.
2. El punto naranja aislado, en forma de círculo, a izquierda entre $100$ y $200$ dolares de costo de pasaje fue una mujer, infante, de segunda clase que no sobrevivió al hundimiento.
3. Los dos puntos azules, casi en la mitad del gráfico, entre $200$ y $300$ dolares de precio, fueron dos hombres de segunda clase que no sobrevivieron y que pagaron un precio exágerado por subir al barco.
4. El punto naranja más arriba de la gráfica pagó más de $500$ dolares por subir al barco, fue mujer de segunda clase y sobrevivió al accidente.
5. Muy al externo derecho y abajo de la figura, hay un punto azul muy pequeño en forma de cículo. Esta persona fue un hombre anciano de tercera clase que murió en el accidente.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Datos
df = pd.read_csv('Libros.csv')
```

```python
#Gráficos
fig, ax = plt.subplots(1,1, figsize=(4,4))

ax = sns.histplot(data=df, x='Price', bins=20, hue='Genre', stat='percent', multiple='dodge')
ax.set_xlabel('Precio')
ax.set_ylabel('Porcentaje')
ax.set_yticks([0,5,10,15,20])
ax.legend(['Ficción','No Ficción'], title='Género', loc='upper right')

fig.show()
```

Del diagrama se puede concluir que:

1. Los libros de no ficción son los más vendidos.
2. El rango de precio de los libros oscila entre $0$ y $30$ dolares.
3. En general, los libros de ficción son más baratos que los de no ficción.
