¿Cuáles son las diferencias entre el operador $+$, el método $add$ y una función que implemente ciclos para sumar los elementos de dos listas?

***Ayuda 1:*** Use los resultados del ejercicio resuelto 6.6.   
***Ayuda 2:*** Use el comando mágico $\%timeit$ para medir el tiempo.

```python
import numpy as np
```

```python
def suma(l1,l2):
  suma = []
  for i,j in zip(l1,l2):
    suma.append(i+j)
  return suma
```

```python
numeros_1 = np.linspace(0,1,10000)
numeros_2 = np.linspace(-1,0,10000)

%timeit numeros_1 + numeros_2
%timeit np.add(numeros_1, numeros_2)
%timeit suma(numeros_1, numeros_2)
```

```python
import numpy as np
```

```python
def contar(lista,caracter,repeticion):
  enmascarar = []

  for cadena in lista:
    if cadena.count(caracter)==repeticion:
       enmascarar.append(True)
    else:
      enmascarar.append(False)

  return enmascarar
```

## <font color='green'>**Ejercicio 39** </font>

Generalice el ejercicio propuesto 9.38 para incluir nombres que:

1. Contengan exactamente dos $a$.
2. Contengan al menos dos $a$.
3. Contengan a lo sumo dos $a$.




```python
import numpy as np
```

```python
def contar(lista, caracter, repeticion, operacion):
  enmascarar = []

  for cadena in lista:
    condicion = eval(str(cadena.count(caracter)) + operacion + str(repeticion))
    if condicion:
       enmascarar.append(True)
    else:
      enmascarar.append(False)

  return enmascarar
```

```python
nombres = np.array(['maria','ana','camilo','alejandra'])
cond_1 = contar(nombres,'a',2,'==')
cond_2 = contar(nombres,'a',2,'>=')
cond_3 = contar(nombres,'a',2,'<=')

print(f'          Dos "a": {nombres[cond_1]}')
print(f' Al menos dos "a": {nombres[cond_2]}')
print(f'A lo sumo dos "a": {nombres[cond_3]}')
```
