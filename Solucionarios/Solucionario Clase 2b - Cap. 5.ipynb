## <font color='green'>**Ejercicio 6** </font>

Confeccione un código que lleve a cabo un descuento, tal que:

1. Si el precio es mayor a $10.000$, entonces el descuento es del $20\%$.
2. Si el precio está entre $5.000$ y $10.000$ (incluyendo), entonces el descuento es del $10\%$.
3. Si el precio es inferior a $5.000$, entonces no hay descuento.

***Ayuda 1:*** Use los resultados del ejercicio propuesto 2.17.   
***Ayuda 2:*** Puede que le sean útiles los condicionales $if$, $elif$ y $else$.

```python
p = int(input('Precio: '))

if p>10000:
  r = p-p*(20/100)
elif p>=5000 and p<=10000:
  r = p-p*(10/100)
else:
  r = p

print(f'Precio con descuento: {r:.1f}')
```
