***Ayuda:*** Puede que le sea útil el método $replace$.

```python
def quitar_tildes(palabra):
    vocales = [['á','a'],['é','e'],['í','i'],['ó','o'],['ú','u']]
    for a,b in vocales:
        palabra = palabra.replace(a,b)
    return palabra

quitar_tildes('polación')
```

## <font color='green'>**Ejercicio 8** </font>

Codifique una función que para de una lista de números retorne:

1. El número par más grande.
2. El número impar más pequeño.


```python
def func(lista):
  par = []
  impar = []

  for n in lista:
    if n%2==0: par.append(n)
    else: impar.append(n)

  n1 = max(par)
  n2 = min(impar)
  return n1,n2

func([1,5,-4,0,-3,2])
```

## <font color='green'>**Ejercicio 9** </font>

Cree una función que elimine los elementos repetidos de una lista.

***Ayuda:*** No use la función nativa $set$.


```python
def eliminar_repetidos(lista):
  lista_sr = []
  for n in lista:
    if n not in lista_sr:
      lista_sr.append(n)
  return lista_sr

eliminar_repetidos([1,2,2,3,3,3,4,5,5])
```

## <font color='green'>**Ejercicio 10** </font>

Escriba una función que le permita calcular el perímetro y el área de un rectángulo de lados $a$ y $b$. Haga esto para la lista de rectángulos:

&emsp; $rectangulos = [(1,1),(1,2),(2,2),(2,3)]$

***Ayuda:*** Perímetro $P=2(a+b)$ y área $A=ab$.

```python
rectangulos = [(1,1),(1,2),(2,2),(2,3)]

def perimetro_area(lista):
  for a,b in lista:
    P = 2*(a + b)
    A = a * b
    print(f'Perímetro: {P:4.1f} -  Área: {A:.1f}')

perimetro_area(rectangulos)
```

***Ayuda:*** Perímetro $P=2\pi r$ y área $A=\pi r^2$.

## <font color='green'>**Ejercicio 12** </font>

Cree una función que reciba una cadena de caracteres como argumento y devuelva el carácter más común.

```python
def caracter_mas_comun(cadena):
  conteo = {}
  for letra in cadena:
    if letra not in conteo:
      conteo[letra] = 1
    else:
      conteo[letra] += 1

  mayor = max(conteo.values())
  for letra in conteo:
    if conteo[letra]==mayor:
      print(letra, end=' ')

caracter_mas_comun('Holaaa Mundoo')
```

## <font color='green'>**Ejercicio 13** </font>

Escriba una función que reciba una fecha como argumento y verifique si es una fecha válida y si está en el formato DD/MM/AAAA.

```python
def verificar_fecha(fecha):
  fecha = fecha.split('/')
  D = fecha[0]; M = fecha[1]; A = fecha[2]
  meses = {'01':31,'02':28,'03':31,'04':30,'05':31,'06':30,'07':31,'08':31,'09':30,'10':31,'11':30,'12':31}

  if len(D)==2 and len(M)==2 and len(A)==4:
    if M in meses.keys():
      if int(D)<=meses[M]:
        return True
      else:
        return False
    else:
      return False
  else:
    return False

verificar_fecha('30/09/2022')
```
