### **Punto #6** Cree una columna para el nombre de la revista y otra para la extensión del dominio. Llénelas como corresponde.

```python
# Instalo una libreria que busque que es util para extraer dominios y extension de dominio.

!pip install tldextract
```

```python
import tldextract # importo la libreria

def Revista(x):
  val = tldextract.extract(x) # sirve para extrar info de una URL
  dominio = val.domain # se toma el dominio
  extension = val.suffix # se toma la extesnion
  return [dominio, extension] # Retorna una lista con el dominio y la extension

oper = list(map(Revista, df['URL'])) # se hace una funcion map que opera sobre cada fila de df['URL'] la funcion revista


# Se crean dos listas vacias
dominio = []
extension = []

for i in range(0,len(df['URL'])): # Se hace un for para que recorra el # de filas del dataFrame
  dominio.append(oper[i][0]) # se agrega el primer valor al dominio
  extension.append(oper[i][1]) # se agrega el segundo valor de la lista a la extension
```

```python
df['Nombre_Revista'] = dominio # se agrega una nueva columna con la info del nombre de la revista
df['Extension'] = extension # se agrega una nueva columna con la info de la extension
```

### **Punto #9**  Cree un archivo Excel donde guarde la información en diferentes hojas según cada revista.

```python
!pip install xlsxwriter
```

```python
# Crear un escritor de Excel
writer = pd.ExcelWriter('informacion_revistas.xlsx', engine='xlsxwriter')

# Obtener la lista de revistas únicas en el DataFrame
revistas = df['Nombre_Revista'].unique()

# Iterar sobre las revistas y guardar cada una en una hoja separada
for revista in revistas:
    df_revista = df[df['Nombre_Revista'] == revista]
    df_revista.to_excel(writer, sheet_name=revista, index=False)

# Guardar el archivo Excel
writer.save()
```
