## <font color='green'>**Ejercicio 2** </font>

Implemente una función para encontrar el mínimo número en una lista.

***Ayuda:*** No use las funciones nativas $min$ y $max$.



````python
#@title Punto2-parte1
import random # Se importa la libreria random
val_ini = 0
val_fini = int(input("Ingrese el tamano que quiere en la list: "))
lista0 = [random.randint(1,100) for i in range(val_ini,val_fini)] # Se genera una comprenhension list para generar numeros aleatorias entre 1 y 100 y que genere 10 valores
print(f'La lista es {lista0}')
````

````python
#@title Punto2-Parte 2
'''
  Voy a utilizar el algoritmo conocido como bubble sort algorithm.
'''
def bubbleSort(lista) -> int:
  """
  Esta funcion permite ordenar una lista de cualquier dimension.
  Adicionalmente, se queda con el primer elemento de la lista que seria el menor.

  Params:
  -------------------------------------------------------------
  lista: type list en python
  """
  n = len(lista) # Longitud de la lista
  for i in range(n): # Primer for que arran en 1 y va hasta n
      for j in range(0, n - i - 1): # Va disminuyendo el tamano, es decir compara entre ultimos elementos
          if lista[j] > lista [j+1]:
              lista[j], lista[j+1] = lista[j+1], lista[j] # Permite intercambiar elementos

  return lista
lista1 = bubbleSort(lista0)
print(f'La lista ordenada es \n{lista1}')
print(f'El valor minimo es\n{lista1[0]}')
````

## <font color='green'>**Ejercicio 4** </font>

Cree una función que elimine los elementos repetidos de una lista.

***Ayuda:*** No use la función nativa $set$.

````python
#@title Punto 4 - Opcion 1
"""
  Este Algoritmo es bueno, pero estoy ordenando una lista para poder eliminar sus valores repetidos
  ¿Qué pasa si la lista debe mantenerse desordenada y eliminar valores de esta?
"""

l2 = [random.randint(1,10) for i in range(0,10)] # Se genera una comprenhension list
print(f'La lista inicial es: {l2}')

l2 = sorted(l2) # por facilidad primero ponemos la lista en orden
print(f'La lista ordenada es: {l2}')

new_list = []
k = 0
while k < len(l2) - 1:
  if l2[k] != l2[k+1]:
    new_list.append(l2[k])
  k += 1

if new_list[-2] != l2[-1]:
  new_list.append(l2[-1])

print(f'La lista sin valores repetidos es: {new_list}')
````

### <font color='green'>**Ejercicio Propuesto 8** </font>

Escriba un programa que le permita calcular el perímetro y el área de un tríangulo equilatero de lado $a$.

***Ayuda:*** Perímetro $P=3a$ y área $A=\frac{\sqrt{3}}{4}a^2$.


````python
lado = float(input("Ingrese el lado del triangulo"))

def PA(a):
  perimetro = 3*a
  area = 3**(1/2) * a**2/4
  return perimetro, area

per, area = PA(lado)
print(f"El lado ingresado es {lado}, el triangulo tiene un perimetro de: {per} y una area de {area}")
````
