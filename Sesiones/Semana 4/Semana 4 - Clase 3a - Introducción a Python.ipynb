## **8.2. Funciones**

Una función permite definir un bloque de código para ser reutilizable dentro del programa. Su sintaxis es:

```
def nombre_función(argumentos):
  <<Ejecutar este bloque de código>>
  return <<Valor/es a devolver>>
```

```
def nombre_función(argumentos):
  <<Ejecutar este bloque de código>>
```

En general, es necesario usar `return` al final para que la función sepa exactamente qué debe devolver, sin embargo, no es obligatorio. Veamos:


````python
def primera_funcion():
  return print('Hola Mundo')

primera_funcion()
````

````python
def primera_funcion():
  print('Hola Mundo')

primera_funcion()
````

````python
def sumar(a,b):
  suma = a+b
  return suma

sumar(2,5)
````

````python
def sumar(a,b):
  suma = a+b

sumar(2,5)
````

Es importante destacar, que no es posible acceder a las variables inicializadas dentro de una función:

````python
suma
````

Además, una función puede devolver cualquier cantidad de resultados:

````python
def funcion_1(a,b):
  return a+b

def funcion_2(a,b):
  return a+b,a-b,a*b,a/b

print(funcion_1(2,2))
print(funcion_2(2,2))
````

Por otro lado, la sentencia `pass` se usa cuando una función no está implementada, pero se quiere llevar a cabo en el futuro. Veamos:

````python
def sumar(a,b):
````

````python
def sumar(a,b):
  pass
````

En conclusión, las funciones son una forma hacer reutilizable un código cuando éste se usa de manera muy repetitiva dentro de un programa. Además, si se comente un error al escribirla y luego se corrige el código correspondiente, automáticamente se enmenda el fallo en cada lugar donde se ha hecho uso de la función.

## **8.3. Funciones Nativas**

Por defecto, Python viene con funciones que agregan versatilidad y robustes a los programas. Algunas de ellas son:

<center>

|Función|Descripción|
|:-:|-|
| input()                  | Lee una línea de entrada por teclado como un carácter. |
| print()                  | Imprime una línea de salida en pantalla. |
| len(objeto)              | Retorna el tamaño de un objeto. |
| round(numero,cifras)     | Retorna un número redondeado. |
| range(inicio,fin,paso)   | Devuelve una secuencia de números. |
| min(iterable)            | Retorna el menor objeto de un iterable. |
| max(iterable)            | Retorna el mayor objeto de un iterable. |
| int(objeto)              | Convierte a número entero. En caso de que se pueda. |
| float(objeto)            | Convierte a número decimal. En caso de que se pueda. |
| str(objeto)              | Convierte a cadena de caracteres. En caso de que se pueda. |
| bool(objeto)             | Convierte a buleano. En caso de que se pueda. |
| list()                   | Crea una lista. |
| dict()                   | Crea un diccionario. |
| truple()                 | Crea una trupla. |
| set()                    | Crea un conjunto. |
| map(funcion,iterable)    | Aplica una función a todos los elementos en una lista. |
| filter(funcion,iterable) | Extrae elementos de un iterable según una función. |
| sorted(iterable,reverse) | Retorna una lista ordenada de un iterable. |
| type(objeto)             | Devuelve el tipo de los objetos. |
| zip(iterable1,iterable2) | Empareja los elementos de dos iterables. |
| enumerate(iterable)      | Agrega un contador a un iterable. |

</center>

Veamos algunos ejemplos:

````python
#Round
round(5.451,1)
````

````python
#Min and Max
lista = [1,0,7,10,9]
print(min(lista), max(lista))
````

````python
#Filter
lista = [1,0,7,10,9]
def par(n): return n%2==0

#filter(par, lista)
list(filter(par, lista))
````

````python
#Map
lista = [1,0,7,10,9]
def suma(n): return n+2

#map(par, lista)
list(map(suma, lista))
````

````python
#Sorted
lista = [1,0,7,10,9]
sorted(lista)
````
