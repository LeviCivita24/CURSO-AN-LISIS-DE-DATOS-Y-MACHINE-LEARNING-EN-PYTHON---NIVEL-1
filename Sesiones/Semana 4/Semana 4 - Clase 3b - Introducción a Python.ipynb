## **8.5. Funciones Anónimas**

Las funciones anónimas sirven para crear pequeñas funciones sin nombre y en una sola línea. Su sintaxis es:

```
lambda argumentos: operaciones
```

Veamos:

````python
#Estándar
def suma(a,b):
  return a+b

print(suma)
print(suma(1,2))
````

````python
#Anónima
suma = lambda a,b: a+b

print(suma)
print(suma(1,2))
````

Nótese que la función creada por la forma estándar tiene como nombre `suma`, mientras que la función anónima lleva como nombre `<lambda>`, de ahí su denominación de sin nombre.
