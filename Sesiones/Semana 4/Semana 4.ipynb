```python
palabra = "Hooolllaa"
letras = []

#Obtener las letras de la palabra
for letra in palabra: letras.append(letra)
letras_sr = list(set(letras)) #{'H', 'a', 'l', 'o'} --> ['H', 'a', 'l', 'o']

#Contar cuantas veces se repite cada letra
contador = []
for letra in letras_sr: #['H', 'a', 'l', 'o']
  n = 0

  for i in palabra: #"Holaa"
    if letra==i: #Coincidencia
      n = n+1

  contador.append(n)

aux = 0 #POS del mayor en contador
for j in range(len(contador)): #Posiciones de contador: [0,1,2,3]
  print(j, contador[j], aux)
  if contador[j]>contador[aux]:
    aux = j

contador, aux, letras_sr[aux]
```

```python
from collections import Counter

Counter(letras)
```

```python
def caracter_mas_comun(cadena):
  conteo = {}
  for letra in cadena:
    if letra not in conteo:
      conteo[letra] = 1
    else:
      conteo[letra] += 1

  mayor = max(conteo.values())
  for letra in conteo:
    if conteo[letra]==mayor:
      print(letra, end=' ')

caracter_mas_comun('"Hooolllaa"')
```

```python
from collections import Counter

def caracter_mas_comun(cadena):
    contador = Counter(cadena)
    carac_mas_comun = contador.most_common(1)[0][0]
    return carac_mas_comun

# Ejemplo de uso
texto = "Hooolllaa"
resultado = caracter_mas_comun(texto)
print(f"El carácter más común es: '{resultado}'")
```

Escriba una función que reciba una fecha como argumento y verifique si es una fecha válida y si está en el formato DD/MM/AAAA.

```python
meses = {'01':31,'02':28,'03':31,'04':30,'05':31,'06':30,
         '07':31,'08':31,'09':30,'10':31,'11':30,'12':31}

meses['08']
#nom_meses = ['01','02',...]
#dias_meses = [31,28,...]
#[['01',31],[]]
```

```python
meses.keys()
```

```python
fecha = '10/08/2023'

def verificar_fecha(cadena):
  lista = cadena.split('/') #lista = ['10','08','2023']
  dia = lista[0]; mes = lista[1]; año = lista[2]
  meses = {'01':31,'02':28,'03':31,'04':30,'05':31,'06':30,
           '07':31,'08':31,'09':30,'10':31,'11':30,'12':31}

  if len(dia)==2 and len(mes)==2 and len(año)==4: #Formato Correcto
    if mes in meses.keys(): #El mes está correcto
      if 0<int(dia)<=meses[mes]: #10<=meses['08'] (10<=31)
        print('Fecha y formato correcto.')
      else:
        print('El día sobrepasa los días del mes.', f'Día es {dia} y el mes tiene {meses[mes]} días.')
    else:
      print('El mes está incorrecto.', mes)
  else:
    print('Error de formato.', cadena)

verificar_fecha(fecha)
```
