A partir del ejercicio propuesto 11.16, efectúe y responda:

1. Filtre aquellos equipos que tengan entre $1$ y $3$ campeonatos.
2. Filtre aquellos equipos que han pasado a más de $15$ mundiales y juegan en Europa (son de la $UEFA$).
3. ¿Cuáles son los equipos que han perdido $2$ o más finales?
4. ¿Cuáles son los equipos que han jugado finales pero no han quedado campeones?

***Ayuda:*** Puede que le sea útil el método $between$.

```python
import pandas as pd
```

```python
df = pd.read_excel('ParticipacionesMundiales.xlsx')
df.sample(5)
```

1. Filtre aquellos equipos que tengan entre 1 y 3 campeonatos.

```python
cond = df.loc[:, 'Títulos'].between(1,3) #Índices de <<aquellos>> que cumple la condición.
df.loc[cond, :]
```

2. Filtre aquellos equipos que han pasado a más de 15 mundiales y juegan en Europa (son de la UEFA).   

* df.loc[:, ''] : Permite tomar una columna en especial.
* df.loc[cond, :] : Permite tomar unos índices específicos y todas las columnas.

```python
cond = (df.loc[:, 'Primera Ronda']>15) & (df.loc[:, 'Confederación']=='UEFA') #and:& o or:|
df.loc[cond, :]
```

3. ¿Cuáles son los equipos que han perdido 2 o más finales?

```python
resta = df.loc[:, 'Finales'] - df.loc[:, 'Títulos']
df.insert(3, 'Finales Perdidas', resta)
df.head(5)
```

```python
cond = df.loc[:, 'Finales Perdidas'] >= 2
df.loc[cond, :]
```

4. ¿Cuáles son los equipos que han jugado finales pero no han quedado campeones?

```python
cond = (df.loc[:, 'Finales']!=0) & (df.loc[:, 'Títulos']==0)
df.loc[cond, :]
```

5. Países que sean de Europa o comiencen por P.

```python
cond1 = (df.loc[:, 'Confederación']=='UEFA')
cond2 = (df.loc[:, 'Selección'].str.replace(u'\xa0','').str.startswith('P'))
cond = cond1 | cond2
df.loc[cond, :]
```
