```python
#Crear/Importar Funciones (funciones.py --> from funciones import quitar_tilde)
def quitar_tilde(palabra):
  vct = ["á","é","í","ó","ú","Á","É","Í","Ó","Ú"]
  vst = ["a","e","i","o","u","A","E","I","O","U"]

  for i in range(10):
    palabra_limpia = palabra.replace(vct[i],vst[i])
    palabra = palabra_limpia

  return palabra_limpia
```

¿Qué sucede si un nombre se repite? ¿qué hacemos con el correo?

```python
def crear_correo(nombre, col_nombre):
  nombre_ = nombre.lower()
  nombre_ = nombre_.replace(' ','.')
  nombre_ = quitar_tilde(nombre_)
  correo  = nombre_ + '@udea.edu.co'

  if nombre in col_nombre.values: #Nombre Existe
    rep = col_nombre.value_counts() #Cuantas veces se repite cada nombre
    if rep[nombre]>=1: #Nombre Repetido
      return nombre_ + str(rep[nombre]) + '@udea.edu.co'
  else: #Nombre No Existe == Nombre Sin Repetir
    return correo

def agregar_datos(tabla):
  nombre = input('Ingrese nombre y apellido: ')
  puntaje = float(input('Ingrese puntaje: '))
  nota = round(puntaje*5/30, 2)
  correo = crear_correo(nombre, tabla.loc[:, 'Nombre'])
  aprobar = 'NO' if nota<3.0 else 'SÍ'

  i = tabla.shape[0] #Índice (el último)
  tabla.loc[i, :] = [nombre, puntaje, nota, correo, aprobar]
```
