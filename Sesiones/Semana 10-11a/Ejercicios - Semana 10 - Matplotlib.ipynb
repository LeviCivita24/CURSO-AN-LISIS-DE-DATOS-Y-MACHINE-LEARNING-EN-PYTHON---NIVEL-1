### <font color='green'>**Ejercicio 10** </font>

Traze en la misma gráfica las siguientes funciones:

&emsp;&emsp;&emsp; $y_1=x+1$,  
&emsp;&emsp;&emsp; $y_2=x^2+2$,   
&emsp;&emsp;&emsp; $y_3=x^3+3$,

para valores de $x \in [-1,1]$, teniendo en cuenta que:

1. Los marcadores del eje $x$ deben ser $-1$, $-1/2$, $0$, $1/2$, $1$.
2. Los marcadores del eje $y$ deben valores enteros.
3. Los ejes deben estar nombrados con las etiquetas $x$ y $f(x)$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100)
y1 = x+1
y2 = x**2+2
y3 = x**3+3
```

```python
fig, ax = plt.subplots(1,1, figsize=(3,3))

ax.plot(x,y1, color='y')
ax.plot(x,y2, color='b')
ax.plot(x,y3, color='g')
ax.axvline(0.5, color='cyan', linestyle='-.')
ax.axhline(0.5, color='cyan', linestyle='-.')
ax.text(-0.75,1.0, 'Hola', color='red', rotation=45)
ax.set_xticks([-1,-1/2,0,1/2,1])
ax.set_xticklabels(['$-1$','$-\\frac{1}{2}$','$0$','$1/2$','$1$'])
ax.set_xlabel('$\\alpha$')
ax.set_ylabel('$f(\\alpha)$')

fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(t): return np.cos(2*np.pi*t) * np.exp(-t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
fig = plt.figure(figsize=(8,4))

#1ra
ax = fig.add_subplot(1,2,1)
ax.plot(t2, f(t2), color='k', linestyle='--')
ax.plot(t1, f(t1), color='k', marker='o', markerfacecolor='green')
ax.grid(True)
ax.set_xlabel('Hola')

#2da
ax = fig.add_subplot(1,2,2, projection='3d')
ax.plot_surface(X,Y,Z)
ax.set_zlim(-1,1)
ax.set_xlabel('x')
ax.set_ylabel('y')

fig.supxlabel('Year', y=-0.05)
fig.show()
```
