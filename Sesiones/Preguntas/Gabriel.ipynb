Escriba un programa que pida al usuario tres números y calcule el promedio aritmético de estos. El resultado debe contener dos decimales. Use ciclos para ello. Note que el resultado es idéntico al problema 7, pero la forma de llevarlo a cabo es diferente.
Profe antes de finalizar sáqueme de la duda con ese ejercicio

```python
n = 3
suma = 0

for i in range(n):
  num = float(input('Ingrese número: '))
  suma = suma + num

promedio = suma/n
print(promedio)
```
