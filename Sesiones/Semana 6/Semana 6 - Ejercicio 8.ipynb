10. Reemplace los valores de la columna Aprueba que contiene los valores Yes y Not, por los valores Sí y No.


```
#Forma 1
apr_trad = []

for palabra in df.loc[:, 'Aprueba']:
  if palabra=='Yes':
    apr_trad.append('Sí')
  else:
    apr_trad.append('No')    

df.loc[:, 'Aprueba'] = apr_trad
```



````python
#Forma 2
cond = df.loc[:, 'Aprueba'] == 'Yes' #Condición (Lista de índices con True/False)
df.loc[cond, 'Aprueba'] = 'Sí' #Ir a la parte de la tabla donde en Aprueba hay la cadena 'Yes'
df.loc[~cond, 'Aprueba'] = 'No' #Ir a la parte de la tabla donde en Aprueba hay la cadena 'Not'
df
````
