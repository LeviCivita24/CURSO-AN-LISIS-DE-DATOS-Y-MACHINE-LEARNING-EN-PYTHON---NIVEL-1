### **9.6.1. Indexación y Segmentación**

Los Arreglos de NumPy, tienen la misma semántica de indexación y segmentación que las listas de Python Nativo. Veamos:

```python
a = np.array([1,5,7,8,9,4,7,2])
a
```

```python
#Indexación
a[3]
```

```python
#Segmentación
a[2:6]
```

```python
#Indexación Sofisticada
a[[1,3,4]]
```

Además:

```python
a = np.arange(16).reshape(4,4)
a
```

```python
#Primera Fila
a[0,:]
```

```python
#Primera Columna
a[:,0]
```

```python
#Segmentar Filas Pares
a[::2,:]
```

```python
#Segmentar Columnas Impares
a[:,1::2]
```

```python
#Segmentar Matriz Interna 2x2
a[1:3,1:3]
```

## **9.7. Números Aleatorios**

NumPy contiene funciones para la generación de números aleatorios. Algunas son:

<center>

| Método | Descripción |
|-:|:-|
| randint(min,max,size)    | Genera números aleatorios de tipo entero. |
| randf(size)              | Genera números aleatorios de tipo flotante entre [0.0,1.0). |
| choice(Arreglo)            | Genera una muestra aleatoria dado un Arreglo. |
| permutation(Arreglo)       | Genera una permutación aleatoria dado un Arreglo. |
| uniform(min,max,size)    | Genera números aleatorios siguiendo una distribución uniforme entre [min,max). |
| normal(mean,std,size)    | Genera números aleatorios siguiendo una distribución normal (Gaussiana). |

</center>

Su sintaxis es `np.random.funcion()`. Veamos :



```python
a = np.array([1,2,3,4,5])
a
```

```python
#Enteros, Flotantes, Escogencia y Permutación
print('     Entero:', np.random.randint(0,5,5))
print('   Flotante:', np.random.ranf(5))
print(' Escogencia:', np.random.choice(a))
print('Permutación:', np.random.permutation(a))
```

```python
#Distribuciones
print('Uniforme:', np.random.uniform(-1,1,5))
print('  Normal:', np.random.normal(0,1,5))
```

Finalmente, con la función `seed` se puede establecer la semillia que generará los números aleatorios. Veamos:

```python
#Sin Seed
np.random.uniform(-1,1,5)
```

```python
#Seed 1
np.random.seed(1)
np.random.uniform(-1,1,5)
```

```python
#Seed 58
np.random.seed(58)
np.random.uniform(-1,1,5)
```

Para más información visite [Random Generator](https://numpy.org/doc/stable/reference/random/generator.html) y [Random Sampling](https://numpy.org/doc/1.16/reference/routines.random.html).
