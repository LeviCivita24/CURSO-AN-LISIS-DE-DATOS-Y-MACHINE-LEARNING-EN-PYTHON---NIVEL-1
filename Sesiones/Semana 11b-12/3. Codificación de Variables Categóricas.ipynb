**Codificación de Variables Categóricas**

```python
#Ordinal
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
#Nominal
from sklearn.preprocessing import OneHotEncoder

import pandas as pd
import numpy as np
```

```python
#Datos
servicio = ['Muy Insatisfecho', 'Insatisfecho', 'Neutral', 'Satisfecho', 'Muy Satisfecho'] #ORDEN
calidad  = ['Mala', 'Buena', 'Muy Buena', 'Excelente'] #ORDEN

encuesta = {' Servicio' : ['Muy Insatisfecho', 'Insatisfecho', 'Neutral', 'Satisfecho', 'Muy Satisfecho','Muy Insatisfecho'],
            'Alimentos' : ['Mala', 'Buena', 'Muy Buena', 'Excelente', 'Mala', 'Buena'],
            'Cliente Frecuente' : ['No', 'No', 'Sí', 'Sí', 'No', 'Sí']}

cliente = ['No', 'No', 'Sí', 'Sí', 'No', 'Sí'] #Cliente Frecuente

pd.DataFrame(encuesta)
```

**1. Codificación Ordinal - 1,2,3,4,...:**

```python
servicio
```

```python
calidad
```

```python
#OrdinalEncoder - MATRICES/TABLAS
datos_ord = pd.DataFrame(encuesta) #Crear el DF
codificador = OrdinalEncoder(categories=[servicio, calidad]) #OBJ. CODIFICAR EN ORDEN 0,1,2,3,... #[], {}, DF, 'blabla..'
codificacion = codificador.fit_transform(datos_ord)

datos_ord = pd.DataFrame(codificacion, columns=['Servicio','Alimentos'])
datos_ord
```

```python
print(codificador.categories_)
```

```python
[array(['Insatisfecho', 'Muy Insatisfecho', 'Muy Satisfecho', 'Neutral',
       'Satisfecho'], dtype=object), array(['Buena', 'Excelente', 'Mala', 'Muy Buena'], dtype=object)]
```

```python
cliente
```

```python
#LabelEncoder - VECTOR/LISTAS
codificador = LabelEncoder()
codificacion = codificador.fit_transform(cliente)
datos_lab = pd.DataFrame(codificacion, columns=['Cliente'])

datos_lab
```

**2. Codificación Nominal:**

```python
datos_one = pd.DataFrame(encuesta)
codificador = OneHotEncoder() #OBJ. CODIFICAR SIN ORDEN --> CÓDIGO BINARIO
codifcacion = codificador.fit_transform(datos_one).toarray()  #toarray() para convertir la matriz dispersa en una matriz como tal

#DF
datos_one = pd.DataFrame(codifcacion, columns=np.concatenate(codificador.categories_))
datos_one
```

```python
print(codificador.categories_)
```

```python
codifcacion = codificador.fit_transform(datos_one)
print(codifcacion)
```

**3. Regresión Logística:**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

c = np.reshape(datos_lab.values, (1,len(datos_lab)))[0]

print('\n OrdinalEncoder: 0,1,2,3,4')
datos_ord = StandardScaler().fit_transform(datos_ord)
m1 = LogisticRegression().fit(datos_ord, c) #MODELO 1
print(f'    Predicciones: {m1.predict(datos_ord)}')
print(f'Clases Correctas: {c}')
print(m1.predict_proba(datos_ord))

print('\n OneHotEncoder: 010101011010')
m2 = LogisticRegression().fit(datos_one, c) #MODELO 2
print(f'    Predicciones: {m2.predict(datos_one)}')
print(f'Clases Correctas: {c}')
print(m2.predict_proba(datos_one))
```
