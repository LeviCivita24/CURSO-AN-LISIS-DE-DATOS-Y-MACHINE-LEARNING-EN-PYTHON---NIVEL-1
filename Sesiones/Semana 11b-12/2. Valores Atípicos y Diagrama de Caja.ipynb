**Valores Atípicos y Diagrama de Caja**

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.style.use('bmh')
```

**Ejemplo 1:**

```python
#Datos
edades = np.array([4,17,29, 33,50,51,61,71, 85,92,99])
longitud = len(edades)

print(longitud)
print(edades[longitud//2]) #Mediana
print((29+33)/2)
print((71+85)/2)
print(f'{np.mean(edades):0.2f}')
print(f'{np.std(edades):0.2f}')
```

```python
#Gráfico
fig, ax = plt.subplots(1,1, figsize=(4,4))

ax.boxplot(edades)
ax.scatter(np.ones(11), edades)
ax.set_yticks(np.arange(0,105,5))
ax.grid(True)

fig.show()
```

```python
#Estadística
Q0 = np.quantile(edades, 0.00)
Q1 = np.quantile(edades, 0.25) #Cuartíl 1 25%
Q2 = np.quantile(edades, 0.50) #Mediana
Q3 = np.quantile(edades, 0.75) #Cuartíl 3 75%
Q4 = np.quantile(edades, 1.00)
IQR = Q3-Q1
LI = Q1-1.5*IQR
LS = Q3+1.5*IQR

print(f'   Min: {np.min(edades):4.1f}')
print(f'    Q0: {Q0:4.1f}')
print(f'    Q1: {Q1:4.1f}')
print(f'    Q2: {Q2:4.1f}')
print(f'Median: {np.median(edades):4.1f}')
print(f'    Q3: {Q3:4.1f}')
print(f'    Q4: {Q4:4.1f}')
print(f'   Max: {np.max(edades):4.1f}')
print(f'   IQR: {IQR:4.1f}')
print(f'Límites IQR: [{LI:.1f},{LS:.1f}]')
```

```python
edades
```

```python
#Escalamiento, Normalización y Estadarización
edades = pd.DataFrame(edades)

datos_mm = MinMaxScaler().fit_transform(edades)
datos_n  = Normalizer().transform(edades.T).T
datos_ss = StandardScaler().fit_transform(edades)
datos_rs = RobustScaler().fit_transform(edades)

np.set_printoptions(precision=2)
print(f'    MinMax: {datos_mm.T}')
print(f'Normalizer: {datos_n.T}')
print(f'  Standard: {datos_ss.T}')
print(f'    Robust: {datos_rs.T}')
```
