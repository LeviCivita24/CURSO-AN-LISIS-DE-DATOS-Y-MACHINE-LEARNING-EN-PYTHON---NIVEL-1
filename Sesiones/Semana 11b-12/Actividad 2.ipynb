```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

import pandas as pd
import numpy as np
```

```python
#Datos
df = pd.read_csv('Entrenamiento-Prueba.csv')
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
df.head(5)
```

```python
#Datos Faltantes
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
x[:, 1:3] = imputer.fit_transform(x[:, 1:3])
print(x)
```

```python
#Codificar Variables Independientes
from sklearn.preprocessing import RobustScaler
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0]), 
                                     ('num', RobustScaler(), [1,2])])
x = ct.fit_transform(x)
print(x)
```

```python
#Codificar Variables Dependientes
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
#Datos de Entrenamiento y Prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
print(x_train, '\n')
print(x_test, '\n')
print(y_train, '\n')
print(y_test)
```

```python
#Escalado de Caracter√≠sticas
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```

```python
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0]), ('num', RobustScaler(), [1,2])], remainder='passthrough')
ct.fit_transform(x_test)
```
