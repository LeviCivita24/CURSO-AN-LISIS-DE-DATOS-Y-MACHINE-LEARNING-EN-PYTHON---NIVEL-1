```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
#Datos
datos = {'Nombre' : ['Mariana', 'Ana', 'Elsa', 'Gustavo', 'Pedro', 'Raúl', 'Carlos', 'José', 'Luis'],
          'Saldo' : [10000, 8000, 9000, 2000, 2100, 12000, 5000, 10000, 200],
           'País' : ['Argentina', 'Argentina', 'Ecuador', 'Colombia', 'Argentina', 'Colombia', 'México', 'Perú', 'Perú']}

df = pd.DataFrame(datos)
df
```

```python
#Codificación
codificador = OneHotEncoder(sparse=False)
codificacion = codificador.fit_transform(df.loc[:, ['País']])

codificacion
```

```python
#Unir Tablas
nuevas_cols = pd.DataFrame(codificacion, columns=codificador.categories_[0])
df = pd.concat([df.iloc[:, :-1], nuevas_cols], axis=1)
df
```

```python
#Guardar
df.to_excel('Datos.xlsx', index=False)
```
