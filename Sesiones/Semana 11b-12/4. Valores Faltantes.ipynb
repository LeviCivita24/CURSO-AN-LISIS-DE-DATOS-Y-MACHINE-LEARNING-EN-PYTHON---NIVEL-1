**3. K-Vecinos: Regresin**

```python
#Datos
edad = np.array([17,18,18,19,20,21,21,40,41,42,42,43,44,44,45,50,54,60,61,67,67,68,76,79,83])
estudio = np.array([6,6,7,7,8,8,9,10,10,10,11,11,11,11,12,14,16,16,16,16,16,17,19,19,22])
ingreso = np.array([11230,9624,13798,3215,19169,14982,15419,17286,14232,18092,18318,16260,22347,16710,28949,27309,32779,29743,30341,34088,32435,33909,32263,42067,42295])

datos = np.stack((edad, estudio, ingreso), axis=1)
print(datos)
```

```python
#Grficos
fig, ax = plt.subplots(1,1, figsize=(4,4))

ax.scatter(datos.T[0],datos.T[1])
ax.set_xlabel('Edad')
ax.set_ylabel('Estudio')

fig.show()
```

```python
#Escalamiento
escalador = preprocessing.MinMaxScaler()
escalados = escalador.fit_transform(datos)
print(escalados)
```

```python
#Grficos
fig, ax = plt.subplots(1,2, figsize=(8,4))

ax[0].scatter(datos.T[0], datos.T[1])
ax[0].set_xlabel('Edad')
ax[0].set_ylabel('Estudio')
ax[0].set_title('Originales')

ax[1].scatter(escalados.T[0], escalados.T[1])
ax[1].set_xlabel('Edad')
ax[1].set_ylabel('Estudio')
ax[1].set_title('Escalados')

fig.show()
```

```python
#Datos Faltantes
faltantes = escalados.copy()
faltantes[[2,7,12,17,22], 2] = np.nan
print(faltantes)
```

```python
imputer = KNNImputer(n_neighbors=10, weights='uniform')
uni = imputer.fit_transform(faltantes)

imputer = KNNImputer(n_neighbors=10, weights='distance')
dis = imputer.fit_transform(faltantes)

#np.stack((escalados,uni,dis), axis=1)[:, :, 2]
```
