**Escalamiento, Normalización y Estandarización**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.style.use('bmh')
```

```python
#Datos
df = pd.read_csv('Ingresos-Carros.csv', encoding='latin1')
df
```

```python
#Listas
x = df.index
y1 = df.loc[:, 'Ingresos']
y2 = df.loc[:, 'Carros']

print(df.values, '\n')
print(f' Min: {np.min(y1):5d}, {np.min(y2):d}')
print(f' Max: {np.max(y1):5d}, {np.max(y2):d}')
print(f'Mean: {np.mean(y1):5.0f}, {np.mean(y2):.0f}')
print(f' Std: {np.std(y1):5.0f}, {np.std(y2):.0f}')
```

```python
#Visualización de Datos
fig, ax = plt.subplots(1,3, figsize=(12,4))

ax[0].scatter(x,y1, color='blue', s=5)
ax[0].scatter(x,y2, color='brown')
ax[0].set_title('Datos')
ax[0].set_xlabel('Persona')

ax[1].scatter(x,y1, color='blue', s=5)
ax[1].set_title('Ingresos')
ax[1].set_xlabel('Persona')

ax[2].scatter(x,y2, color='brown', s=5)
ax[2].set_title('Carros')
ax[2].set_xlabel('Persona')

fig.show()
```

```python
#Distribución de Datos
fig, ax = plt.subplots(2,2, figsize=(8,8))

ax[0,0].scatter(x,y1, color='blue', s=5)
ax[0,0].set_title('Ingresos')
ax[0,0].set_xlabel('Persona')
ax[0,0].set_xlim([0,10])

ax[0,1].scatter(x,y2, color='brown', s=5)
ax[0,1].set_xlim([0,10])
ax[0,1].set_title('Carros')
ax[0,1].set_xlabel('Persona')

ax[1,0].hist(y1, color='blue', bins=50)
ax[1,0].set_xlabel('Ingresos')

ax[1,1].hist(y2, color='brown', bins=50)
ax[1,1].set_xlabel('Carros')

fig.show()
```

**1. Escalamiento:**

```python
from sklearn.preprocessing import MinMaxScaler

t = MinMaxScaler()
datos_mm = t.fit_transform(df)

print(datos_mm, '\n')
print(f' Min: {np.min(datos_mm[:, 0]):7.4f}, {np.min(datos_mm[:, 1]):7.4f}')
print(f' Max: {np.max(datos_mm[:, 0]):7.4f}, {np.max(datos_mm[:, 1]):7.4f}')
print(f'Mean: {np.mean(datos_mm[:, 0]):7.4f}, {np.mean(datos_mm[:, 1]):7.4f}')
print(f' Std: {np.std(datos_mm[:, 0]):7.4f}, {np.std(datos_mm[:, 1]):7.4f}')
```

```python
#Visualización de Datos
x = df.index
y1 = datos_mm[:,0]
y2 = datos_mm[:,1]
fig, ax = plt.subplots(1,3, figsize=(12,4))

ax[0].scatter(x,y1, color='blue', s=5)
ax[0].scatter(x,y2, color='brown')
ax[0].set_title('Datos')
ax[0].set_xlabel('Persona')

ax[1].scatter(x,y1, color='blue', s=5)
ax[1].set_title('Ingresos')
ax[1].set_xlabel('Persona')

ax[2].scatter(x,y2, color='brown', s=5)
ax[2].set_title('Carros')
ax[2].set_xlabel('Persona')

fig.show()
```

**2. Normalización:**

```python
from sklearn.preprocessing import Normalizer

t = Normalizer()
datos_n = t.fit_transform(df.T).T

print(datos_n, '\n')
print(f' Min: {np.min(datos_n[:, 0]):7.4f}, {np.min(datos_n[:, 1]):7.4f}')
print(f' Max: {np.max(datos_n[:, 0]):7.4f}, {np.max(datos_n[:, 1]):7.4f}')
print(f'Mean: {np.mean(datos_n[:, 0]):7.4f}, {np.mean(datos_n[:, 1]):7.4f}')
print(f' Std: {np.std(datos_n[:, 0]):7.4f}, {np.std(datos_n[:, 1]):7.4f}')
```

**3. Estandarización:**

```python
from sklearn.preprocessing import StandardScaler

t = StandardScaler()
datos_ss = t.fit_transform(df)

print(datos_ss, '\n')
print(f' Min: {np.min(datos_ss[:, 0]):7.4f}, {np.min(datos_ss[:, 1]):7.4f}')
print(f' Max: {np.max(datos_ss[:, 0]):7.4f}, {np.max(datos_ss[:, 1]):7.4f}')
print(f'Mean: {np.mean(datos_ss[:, 0]):7.4f}, {np.mean(datos_ss[:, 1]):7.4f}')
print(f' Std: {np.std(datos_ss[:, 0]):7.4f}, {np.std(datos_ss[:, 1]):7.4f}')
```

```python
from sklearn.preprocessing import RobustScaler

t = RobustScaler()
datos_rs = t.fit_transform(df)

print(datos_rs, '\n')
print(f' Min: {np.min(datos_rs[:, 0]):7.4f}, {np.min(datos_rs[:, 1]):7.4f}')
print(f' Max: {np.max(datos_rs[:, 0]):7.4f}, {np.max(datos_rs[:, 1]):7.4f}')
print(f'Mean: {np.mean(datos_rs[:, 0]):7.4f}, {np.mean(datos_rs[:, 1]):7.4f}')
print(f' Std: {np.std(datos_rs[:, 0]):7.4f}, {np.std(datos_rs[:, 1]):7.4f}')
```

**4. Graficación:**

```python
#Distribución de Datos Transformados - Columna 1
fig, ax = plt.subplots(2,5, figsize=(20,8))

ax[0,0].scatter(x,y1, color='blue', s=5);             ax[0,0].set_title('Original');   ax[0,0].set_xlabel('Persona')
ax[0,1].scatter(x,datos_mm[:, 0], color='blue', s=5); ax[0,1].set_title('MinMax');     ax[0,1].set_xlabel('Persona')
ax[0,2].scatter(x,datos_n[:, 0], color='blue', s=5);  ax[0,2].set_title('Normalizer'); ax[0,2].set_xlabel('Persona')
ax[0,3].scatter(x,datos_ss[:, 0], color='blue', s=5); ax[0,3].set_title('Standar');    ax[0,3].set_xlabel('Persona')
ax[0,4].scatter(x,datos_rs[:, 0], color='blue', s=5); ax[0,4].set_title('Robust');     ax[0,4].set_xlabel('Persona')

ax[1,0].hist(y1, color='blue', bins=50);             ax[1,0].set_xlabel('Ingresos')
ax[1,1].hist(datos_mm[:, 0], color='blue', bins=50); ax[1,1].set_xlabel('Ingresos')
ax[1,2].hist(datos_n[:, 0], color='blue', bins=50);  ax[1,2].set_xlabel('Ingresos')
ax[1,3].hist(datos_ss[:, 0], color='blue', bins=50); ax[1,3].set_xlabel('Ingresos')
ax[1,4].hist(datos_rs[:, 0], color='blue', bins=50); ax[1,4].set_xlabel('Ingresos')

fig.show()
```

```python
#Distribución de Datos Transformados - Columna 2
fig, ax = plt.subplots(2,5, figsize=(20,8))

ax[0,0].scatter(x,y2, color='brown', s=5);             ax[0,0].set_title('Original');   ax[0,0].set_xlabel('Persona')
ax[0,1].scatter(x,datos_mm[:, 1], color='brown', s=5); ax[0,1].set_title('MinMax');     ax[0,1].set_xlabel('Persona')
ax[0,2].scatter(x,datos_n[:, 1], color='brown', s=5);  ax[0,2].set_title('Normalizer'); ax[0,2].set_xlabel('Persona')
ax[0,3].scatter(x,datos_ss[:, 1], color='brown', s=5); ax[0,3].set_title('Standar');    ax[0,3].set_xlabel('Persona')
ax[0,4].scatter(x,datos_rs[:, 1], color='brown', s=5); ax[0,4].set_title('Robust');     ax[0,4].set_xlabel('Persona')

ax[1,0].hist(y2, color='brown', bins=50);             ax[1,0].set_xlabel('Carros')
ax[1,1].hist(datos_mm[:, 1], color='brown', bins=50); ax[1,1].set_xlabel('Carros')
ax[1,2].hist(datos_n[:, 1], color='brown', bins=50);  ax[1,2].set_xlabel('Carros')
ax[1,3].hist(datos_ss[:, 1], color='brown', bins=50); ax[1,3].set_xlabel('Carros')
ax[1,4].hist(datos_rs[:, 1], color='brown', bins=50); ax[1,4].set_xlabel('Carros')

fig.show()
```
