## **10.2. Graficadores**

Matplotlib posee una gran variedad de graficadores, su sintáxis, en general es `graficador(entradas, modificaciones)`. Donde `entradas` son los datos que se quieren graficar y `modificaciones` son los parámetros que permiten llevar a cabo la personalización. Veamos:

<p><a name="m21"></a></p>

### **10.2.1. Básicos**

Entre los gráficadores básicos más comunes se encuentran:

<center>

| Graficador | Descripción |
|-:|:-|
| plot         | Dibujar un diagrama de línea. |
| scatter      | Dibujar un diagrama de puntos (dispersión). |
| bar          | Dibujar un diagrama de barras verticales. |
| barh         | Dibujar un diagrama de barras horizontales. |
| fill_between | Dibujar el área bajo una línea. |

</center>

Veamos algunos ejemplos:

```python
#Datos - Generados con NumPy
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)
```

```python
#Plot
x = np.linspace(0,1,100)
y = np.sin(2*np.pi*x)

#                      F C        Pulgadas
fig, ax = plt.subplots(1,2, figsize=(4,2))
ax[0].plot(x,y)
ax[1].scatter(x,y, s=0.8)
fig.show()
```

```python
#plt.subplot(121)
#plt.plot(x,y)
#plt.subplot(122)
#plt.show()
```

```python
#Scatter
x = np.random.normal(0,1,100) #Datos - Col. 1
y = np.random.normal(0,1,100) #Datos - Col. 2

fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.scatter(x,y)
fig.show()
```

```python
#Bar (Vertical)
x = np.arange(10)
y = np.random.randint(0,5,10)

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.bar(x,y)
fig.show()
```

```python
#Fill Between
np.random.seed(1)
x = np.linspace(0,8,20)
y1 = 3 + 4*x/8 + np.random.uniform(0.0,0.5,20)
y2 = 1 + 2*x/8 + np.random.uniform(0.0,0.5,20)
y = (y1+y2)/2

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.fill_between(x,y1,y2, alpha=0.1)
ax.plot(x,y)
fig.show()
```

Para más información visite [Matplotlib](https://matplotlib.org/stable/api/matplotlib_configuration_api.html) y [Matplotlib Plot Types](https://matplotlib.org/stable/plot_types/index.html).

### **10.2.2. Estadísticos**

Para presentar resultados de índole estadístico, se tienen los graficadores:

<center>

| Gráfica | Descripción |
|-:|:-|
| hist       | Dibujar un histograma (frecuencia de valores). |
| boxplot    | Dibujar un diagrama de cajas y vigotes. |
| errorbar   | Dibujar un diagrama de línea con barra de errores. |
| violinplot | Dibujar un diagrama de violín. |
| hist2d     | Dibujar un histograma bidimensional. |
| pie        | Dibujar un diagrama circular (sectores). |

</center>

Veamos algunos ejemplos:

```python
#Hist
sigma = 1.0
mean = 0.0

#Histograma
x = np.random.normal(mean,sigma,100000)

#Datos - Plot
xx = np.linspace(-3,3,100)
y = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-0.5*(xx-mean)**2/sigma**2)

fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.plot(xx,y, color='red')
ax.hist(x, bins=50, density=True, color='black')
fig.show()
```

```python
x
```

```python
#Boxplot
x = np.random.normal(0,1,1000)

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.boxplot(x)
fig.show()
```

```python
#Errorbar
x = np.linspace(0,1,10)
y = np.sin(2*np.pi*x)
yerr = abs(np.random.normal(0,1,10))

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.errorbar(x,y,yerr)
fig.show()
```

```python
#Violinplot
x = np.random.normal(0,1,1000)

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.violinplot(x)
fig.show()
```

```python
#Hist2D
x = np.random.normal(0,1,1000)
y = np.random.normal(0,1,1000)

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.hist2d(x,y)
fig.show()
```

```python
#Pie
x = [10,15,25,50]

fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.pie(x)
fig.show()
```

Para más información visite [Matplotlib](https://matplotlib.org/stable/api/matplotlib_configuration_api.html) y [Matplotlib Plot Types](https://matplotlib.org/stable/plot_types/index.html).

### **10.2.3. Campos Vectoriales y Contornos**

Para trazar campos vectoriales $(u(x,y), v(x,y))$ y contornos $f(x,y)$, se tienen las siguientes funciones:

<center>

| Gráfica | Descripción |
|-:|:-|
| imshow     | Dibujar un mapa de calor a partir de una matriz. |
| contour    | Dibujar un diagrama de contorno . |
| contourf   | Dibujar un diagrama de contorno con mapa de calor. |
| quiver     | Dibujar un campo vectorial bidimensional. |
| streamplot | Dibujar líneas de corriente de un campo vectorial. |

</center>

Veamos algunos ejemplos:

```python
x, y = np.meshgrid(np.linspace(-2,2,200), np.linspace(-2,2,200))
z = (x**2+y**2)*np.exp(-x**2-y**2)
```

```python
#Imshow
fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.imshow(z, cmap='hsv')
fig.show()
```

```python
#Contour
fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.contour(x,y,z)
fig.show()
```

```python
#Contourf
fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.contourf(x,y,z)
fig.show()
```

```python
x = np.linspace(-4,4,6)
y = np.linspace(-4,4,6)
X, Y = np.meshgrid(x,y)
U = X+Y
V = Y-X
```

```python
#Quiver
fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.quiver(X,Y,U,V)
fig.show()
```

```python
#Streamplot
fig, ax = plt.subplots(1,1, figsize=(2,2))
ax.streamplot(X,Y,U,V)
fig.show()
```

Para más información visite [Matplotlib](https://matplotlib.org/stable/api/matplotlib_configuration_api.html) y [Matplotlib Plot Types](https://matplotlib.org/stable/plot_types/index.html).

### **10.3.1. Colores, Marcadores y Líneas**

Para modificar estos parámetros, dentro del graficador, se usan los comandos:

<center>

| Modificación | Descripción |
|-:|:-|
| color=nombre_color      | Establecer un color para el trazo. |

### **10.3.2. Título, Leyenda y Rejilla**

Para incluir estas características en la figura, se deben llamar los métodos respectivos sobre los ejes creados, así:

<center>

| Método | Descripción |
|-:|:-|
| ax.set_title(modificaciones) | Establecer un título. |
| ax.legend(modificaciones)    | Establecer una leyenda. |
| ax.grid(modificaciones)      | Establecer una rejilla. |

</center>

Veamos algunos ejemplos:

```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
z = np.cos(x)
```

```python
#Title
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y)
ax.plot(x,z)
ax.set_title(label='Funciones Trigonométricas', loc='left',fontsize=8)
fig.show()
```

Para más información visite [Title Method](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html).

```python
#Legend
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y, label='$sin(x)$')
ax.plot(x,z, label='$cos(x)$')
ax.legend(loc='lower left', facecolor='red', edgecolor='k', title='FT')
fig.show()
```

Para más información visite [Legend Method](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html).

```python
#Grid
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y)
ax.plot(x,z)
ax.grid(axis='x', color='r', alpha=0.5)
fig.show()
```

Para más información visite [Grid Method](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html).

### **10.3.3. Ejes**

Para modificar los ejes de la figura, se deben llamar los métodos respectivos sobre los ejes creados, así:

<center>

| Método | Descripción |
|-:|:-|
| ax.set_xlabel(etiqueta)         | Establecer la etiqueta para el eje horizontal. |
| ax.set_ylabel(etiqueta)         | Establecer la etiqueta para el eje vertical. |
| ax.set_xlim([lim-inf, lim-sup]) | Establecer los límites para el eje horizontal. |
| ax.set_ylim([lim-inf, lim-sup]) | Establecer los límites para el eje vertical. |
| ax.set_xticks(marcas)           | Establecer las marcas para el eje horizontal. |
| ax.set_yticks(marcas)           | Establecer las marcas para el eje vertical. |
| ax.set_xticklabels(etiquetas)   | Establecer las etiquetas para las marcas del eje horizontal. |
| ax.set_yticklabels(etiquetas)   | Establecer las etiquetas para las marcas del eje vertical. |
| ax.set_xscale(escala)           | Establecer la escala para el eje horizontal. |
| ax.set_yscale(escala)           | Establecer la escala para el eje vertical. |

</center>

Veamos algunos ejemplos:




```python
x = np.linspace(0,2*np.pi,50)
y = np.sin(x)
```

```python
#Set Label
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y)
ax.set_xlabel('$Eje \ Horizontal$')
ax.set_ylabel('Eje Vertical')
fig.show()
```

```python
#Set Limits
fig, ax = plt.subplots(1,1, figsize=(3,3))
ax.plot(x,y)
ax.set_xlim([-5,10])
ax.set_ylim([0,1])
fig.show()
```

```python
#1D array: Lista - Vectores
#2D array: Tabla - Matrices
#3D array: Libro - Tensor de Rango 3
```

```python
#Set Ticks
fig, ax = plt.subplots(1,1, figsize=(3,3))
ax.plot(x,y)
ax.set_xticks(range(0,18,2)) #Esta función admite listas
#ax.set_xticks([0,1,3,4,7])
#ax.set_xticks([0,np.pi,2*np.pi])
#ax.set_xticklabels(['$0$','$\pi$','$2\pi$'])
ax.set_yticks([-1,-1/2,0,1/2,1])
ax.set_yticklabels(['-1','-1/2','0','1/2','1'])
fig.show()
```

Y si se quiere remover por completo las marcas, entonces:

```python
#Set Ticks (Null)
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y)
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')
fig.show()
```

```python
x = 10.0**np.linspace(0,2,20)
y = x**2
```

```python
#Set Scale
fig, ax = plt.subplots(1,1, figsize=(4,4))
ax.plot(x,y)
ax.set_xscale('log')
ax.set_yscale('log')
fig.show()
```

Para más información sobre las escalas disponibles visite [Scales](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xscale.html).
