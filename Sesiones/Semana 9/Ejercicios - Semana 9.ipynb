### <font color='green'>**Ejercicio 2** </font>

Traze las siguientes funciones:

&emsp;&emsp;&emsp; $y_1=x$,   
&emsp;&emsp;&emsp; $y_2=x/2$,   
&emsp;&emsp;&emsp; $y_3=x^2$,

en una misma gráfica, para valores de $x \in [0,1]$.

***Ayuda:*** Intente usar el graficador `plot` las veces que sean necesarias.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x = np.linspace(0,1,20)
y1 = x
y2 = x/2
y3 = x**2

df = pd.DataFrame(data=[x,y1,y2,y3]).T
df.columns = ['x','y1','y2','y3']
df.sample(5)
```

```python
#Gráfica 1 - Tres conjuntos de datos en un mismo par de ejes
fig, ax = plt.subplots(1,1, figsize=(2,2))

ax.plot(df.loc[:,'x'], df.loc[:,'y1'], color='gold')
ax.plot(df.loc[:,'x'], df.loc[:,'y2'], color='cyan')
ax.plot(df.loc[:,'x'], df.loc[:,'y3'], color='purple')

fig.show()
```

```python
#Gráfica 2  - Tres conjuntos de datos en diferentes par de ejes
fig, ax = plt.subplots(1,3, figsize=(6,2))
fig.subplots_adjust(wspace=0.38)

ax[0].plot(df.loc[:,'x'], df.loc[:,'y1'], color='gold')
ax[1].plot(df.loc[:,'x'], df.loc[:,'y2'], color='cyan')
ax[2].plot(df.loc[:,'x'], df.loc[:,'y3'], color='purple')

fig.show()
```

### <font color='green'>**Ejercicio 6** </font>

Dibuje las siguientes funciones en una misma figura:

&emsp;&emsp;&emsp; $y_1=x$,   
&emsp;&emsp;&emsp; $y_2=x^2$,   
&emsp;&emsp;&emsp; $y_3=\sqrt{x}$,

para $x \in [0,1]$. Las líneas deben quedar en azul, rojo y negro, respectivamente y con diferentes estilos.

***Ayuda:*** Cambie los parámetros `color` y `linestyle` del graficador `plot`.



```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('bmh')
```

```python
#Datos
x = np.linspace(0,1,100)
y1 = x
y2 = x**2
y3 = np.sqrt(x)
```

```python
#Gráfica
fig, ax = plt.subplots(1,1, figsize=(2,2))

ax.plot(x,y1, color='gold', linestyle='-.')
ax.plot(x,y2, color='cyan', linestyle='--')
ax.plot(x,y3, color='purple')

fig.show()
```

### <font color='green'>**Ejercicio 7** </font>

Traze un diagrama de dispersión para los siguientes datos:

&emsp;&emsp; $x = np.linspace(0,10,100)$,   
&emsp;&emsp; $y1 = x + np.random.normal(0,1,100)$,   
&emsp;&emsp; $y2 = 2*x + np.random.normal(0,1,100)$,   
&emsp;&emsp; $y3 = 3*x + np.random.normal(0,1,100)$.

Estos deben quedar representados en tonalidades de color rojo y exhibir diferentes marcadores.

***Ayuda 1:*** Pruebe con los colores `lightcoral`, `brown`, `maroon`.      
***Ayuda 2:*** Pruebe con los marcadores `.`, `1`, `*`.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x = np.linspace(0,10,100)
y1 = x+np.random.normal(0,1,100)
y2 = 2*x+np.random.normal(0,1,100)
y3 = 3*x+np.random.normal(0,1,100)
```

```python
#Gráfica
fig, ax = plt.subplots(1,1, figsize=(2,2))

ax.scatter(x,y1, color='lightcoral', marker='.', s=5, alpha=0.5)
ax.plot(x,x, color='lightcoral', linewidth=1)
ax.scatter(x,y2, color='brown', marker='1', s=5, alpha=0.5)
ax.plot(x,2*x, color='brown', linewidth=1)
ax.scatter(x,y3, color='maroon', marker='*', s=5, alpha=0.5)
ax.plot(x,3*x, color='maroon', linewidth=1)
ax.set_title('Hola', loc='left', fontsize=8)

fig.show()
```

### <font color='green'>**Ejercicio 8** </font>

Genere un gráfico de líneas para el conjunto de datos:

&emsp;&emsp; $x = np.linspace(1,2,100)$,      
&emsp;&emsp; $y1 = x**(-1)$,   
&emsp;&emsp; $y2 = x$,   
&emsp;&emsp; $y3 = x**2$.

El resultado debe contener como título '*Funciones y Potencias*' y las respectivas leyendas.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos
x = np.linspace(1,2,100)
y1 = x**(-1)
y2 = x
y3 = x**2
```

En este ejemplo estamos graficando los datos dados por: $x^{-1}$, $x$ y $x^2$.

```python
#Gráfica
fig, ax = plt.subplots(1,1, figsize=(2,2))

ax.plot(x,y1, color='r', linewidth=1, label='$x^{-1}$')
ax.plot(x,y2, color='b', linewidth=1, label='$x$')
ax.plot(x,y3, color='k', linewidth=1, label='$x^2$')
ax.set_title('Funciones y Potencias', fontsize=8)
ax.legend(facecolor='white', fontsize=8, loc='upper right')

fig.show()
```
