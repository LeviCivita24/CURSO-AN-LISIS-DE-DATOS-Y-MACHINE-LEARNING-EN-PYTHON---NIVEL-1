### **11.8.1. Concat**

El método `concat` sirve para unir DataFrames. Veamos:

```python
datos = {'Nombre': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
        'Puntaje': [12.5, 9, 16.5, 10, 9, 20, 14.5, 12, 8, 19],
       'Intentos': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'Aprueba': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}

indices = ['a','b','c','d','e','f','g','h','i','j']

df = pd.DataFrame(data=datos, index=indices)
df
```

```python
#DataFrame 1
df1 = df.iloc[:3, [0,1]]
df1
```

```python
#DataFrame 2
df2 = df.iloc[1:4, [0,2]]
df2
```

```python
#Por Filas
pd.concat([df1, df2], axis=0, ignore_index=True)
```

```python
#Por Columnas
pd.concat([df1, df2], axis=1, ignore_index=True)
```

Donde se han usado los comandos:

1. `axis=0` para las filas (índices).
2. `axis=1` para las columnas.
3. `ignore_index=True` para reiniciar los índices.



Además, se pueden realizar diferentes operaciones, las cuales serán ilustradas con `merge` y `join`:

### **11.8.2. Merge**

El método `merge` sirve para combinar DataFrames teniendo en cuenta las columnas. Veamos:

```python
#DataFrame 1
df1
```

```python
#DataFrame 2
df2
```

```python
#Inner
pd.merge(df1,df2, how='inner')
```

```python
#Outer
pd.merge(df1,df2, how='outer')
```

```python
#Left
pd.merge(df1,df2, how='left')
```

```python
#Right
pd.merge(df1,df2, how='right')
```

Pandas reconoce que cada DataFrame tiene una columna *name* y los combina referenciando dicha columna. Si hay un dato faltante lo rellena con el valor `NaN` (Not a Number). Para más información visite [Merge](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html).

<p><a name="p83"></a></p>

### **11.8.3. Join**

El método `join` sirve para combinar DataFrames teniendo en cuenta las filas. Veamos:

```python
#DataFrame 1
df1
```

```python
#DataFrame 2
df2.columns = ['Nombre2','Intentos']
df2
```

```python
#Inner
df1.join(df2, how='inner')
```

```python
#Outer
df1.join(df2, how='outer')
```

```python
#Left
df1.join(df2, how='left')
```

```python
#Right
df1.join(df2, how='right')
```

Pandas reconoce que cada DataFrame tiene filas similares y los combina usando dichas filas. Si hay un dato faltante lo rellena con el valor `NaN` (Not a Number). Para más información visite [Join](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html).
