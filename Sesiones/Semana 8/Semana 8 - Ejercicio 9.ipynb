A partir del dataframe del ejercicio resuelto 11.8, efectúe:

1. Cree una tabla que contenga las cuatro primeras filas y las columnas $Nombre$, $Puntaje$ e $Intentos$.
2. Cree otra tabla que contenga las filas desde la $c$ hasta la $f$ y las columnas $Nombre$ y $Aprueba$.
3. Haga una combinación completa de ambos dataframes con el método $merge$.
4. Cambie los valores numéricos que tienen la clausula $NaN$, por el promedio de los que sí tienen valor.
5. Cambie los valores alfabéticos que tienen la clausula $NaN$, por la palabra '$Not$'.

***Ayuda 1:*** Puede que le sea útil el argumento $how$ del método $merge$.         
***Ayuda 2:*** Puede que le sean útiles los métodos $isna$ y $mean$.       


```python
import pandas as pd
```

```python
df = pd.read_excel("Competencia.xlsx")
df.head(5)
```

1. Cree una tabla que contenga las cuatro primeras filas y las columnas Nombre, Puntaje e Intentos.

```python
df1 = df.loc[0:3, ['Nombre','Puntaje','Intentos']]
df1
```

2. Cree otra tabla que contenga las filas desde la 2 hasta la 5 y las columnas Nombre y Aprueba.

```python
df2 = df.loc[2:5, ['Nombre','Aprueba']]
df2
```

3. Haga una combinación completa de ambos dataframes con el método merge.

```python
df3 = pd.merge(df1,df2, how='outer')
df3
```

4. Cambie los valores numéricos que tienen la clausula NaN, por el promedio de los que sí tienen valor.

```python
promedios = df3.mean(numeric_only=True)
prom_puntaje = promedios['Puntaje']
prom_intentos = promedios['Intentos']

print(f"Prom-Puntaje: {prom_puntaje:.2f}")
print(f"Prom-Intentos: {prom_intentos:.2f}")
```

```python
cond1 = df3.loc[:, 'Puntaje'].isna()
df3.loc[cond1, 'Puntaje'] = prom_puntaje
```

```python
cond2 = df3.loc[:, 'Intentos'].isna()
df3.loc[cond2, 'Intentos'] = 1 #prom_intentos
```

5. Cambie los valores alfabéticos que tienen la clausula NaN, por la palabra 'Not'.

```python
cond3 = df3.loc[:, 'Aprueba'].isna()
df3.loc[cond3, 'Aprueba'] = 'Not' #prom_intentos
```
