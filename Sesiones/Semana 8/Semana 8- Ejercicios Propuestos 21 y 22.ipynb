```python
palabras = ['Hello','World','boy','toy']

def traducir(lista):
  lista_trad = []
  traductor = Translator()

  for palabra in lista:
    palabra_trad = traductor.translate(palabra, src='en', dest='es')
    lista_trad.append(palabra_trad.text)

  return lista_trad

traducir(palabras)
```
