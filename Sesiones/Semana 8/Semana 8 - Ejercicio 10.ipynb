```python
from funciones import quitar_tilde
import pandas as pd
```

1. Cree un dataframe con esta base de datos.

```python
df = pd.read_excel('Trabajadores.xlsx', sheet_name='Ventas')
df
```

2. Quíteles las tildes a las palabras de la columna Nombre y deje únicamente las primeras letras en mayúscula.

```python
df.loc[:, 'Nombre'] = df.loc[:, 'Nombre'].apply(quitar_tilde)
df.loc[:, 'Nombre'] = df.loc[:, 'Nombre'].apply(str.title)
```

3. Ponga la primera letra de las palabras de la columna Concepto en mayúscula.

```python
df.loc[:, 'Concepto'] = df.loc[:, 'Concepto'].apply(str.capitalize)
```

4. Separe la tabla en diferentes dataframes, uno por cada nombre.
5. Guarde los dataframes en un achivo de Excel y por cada nombre cree una hoja, incluya la base de datos original.

```python
grupos = list(df.groupby(by='Nombre')) #Formar los grupos
archivo = pd.ExcelWriter('Tabla Final.xlsx') #Archivo de Excel vacío
n = len(set(df.loc[:, 'Nombre'])) #Cantidad de personas

for i in range(n):
  nombre = grupos[i][0]
  tabla = grupos[i][1]
  tabla.to_excel(archivo, sheet_name=nombre, index=False)

archivo.close()
```
