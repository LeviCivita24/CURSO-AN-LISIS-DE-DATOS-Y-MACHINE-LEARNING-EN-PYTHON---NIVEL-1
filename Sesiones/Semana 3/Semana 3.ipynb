```python
#Agregar a grupo
nombres = ['AntoNIA','PAmelA','jUAn','Sherman','BerTA']
sexos = ['M','Mujer','hOmbre','h','MUJER']

grupo_1 = []
grupo_2 = []

for nombre,sexo in zip(nombres,sexos):
  nombre = nombre.lower() #Nombre en minúscula
  sexo_car = sexo[0].capitalize() #Mujer --> M --> M, hOmbre --> h --> H, h --> h --> H

  if (sexo_car=='H' and nombre>'m') or (sexo_car=='M' and nombre<'l'):
    grupo_1.append([nombre.title(),sexo_car])
  else:
    grupo_2.append([nombre.title(),sexo_car])

print(f'Nombres: {nombres} \n')
print(f'Grupo 1: {grupo_1}')
print(f'Grupo 2: {grupo_2}')
```
