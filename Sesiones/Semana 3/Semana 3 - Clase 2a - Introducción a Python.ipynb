### **2.6.1. Indexado**

Para indexar una cadena de caracteres se hace uso de la sintáxis `cadena[posicion]`:



```python
cadena = 'Hola Mundo'

print('Primer Carácter:', cadena[0])
print('Segundo Carácter:', cadena[1])
print('Quinto Carácter:', cadena[5])
```

Donde es importante descatar, que, la indexación siempre comienza en `0` (o primera posición).    

Ahora, si se quiere traer un subconjunto adyacente de caracteres, entonces la sintáxis debe ser `cadena[inicio:fin]`:

```python
print('Parte 1: ', cadena[0:4])
print('Parte 2: ', cadena[4:7])
print('Parte 3: ', cadena[7:10])
```

Nótese nuevamente que no se toma el valor alojado en `fin`, sino el que está en la posición inmediatamente anterior.    

Por otro lado, si no se especifica el `inicio` y el `fin`, se entenderá, por defecto, que se quiere tomar la cadena de caracteres entera:

```python
cadena[:]
```

Pero si se precisa el `fin` pero no el `inicio`, se asignará al `incio` el principio de la cadena de caracteres:

```python
cadena[:3]
```

Y si se especifica el `inicio` pero no el `fin`, se tomará el `fin` como el último índice de la cadena de caracteres:

```python
cadena[6:]
```

Por otro lado, si se quiere por ejemplo saltar caracteres, la síntaxis debe ser `cadena[inicio:fin:paso]`:

```python
#Cada 2
cadena[0:10:2]
```

```python
#Cada 3
cadena[0:10:3]
```

**Nota:** Para los índices hay que tener presente que:

* Cualquier número entero se puede usar como índice.
* Si intenta leer un carácter que no existe, obtendrá un `IndexError` (Error de Indexado).
* Si un índice tiene un valor negativo, cuenta hacia atrás desde el final de la lista.

### **2.6.2. Métodos**

Las cadenas de caracteres son objetos en Python, por consiguiente, poseen métodos (funcionalidades) que operan sobre ellos. Veamos:   

<center>

| Método | Descripción |
|-:|:-|
| cadena.capitalize() | Convertir el primer carácter en mayúscula. |
| cadena.upper()      | Convertir una cadena de caracteres en mayúsculas. |
| cadena.lower()      | Convertir una cadena de caracteres en minúsculas. |
| cadena.title()      | Convertir el primer carácter de cada palabra de la cadena en mayúsculas. |
|||
| cadena.isalnum()   | Devuelve `True` si todos los caracteres de la cadena son alfanuméricos. |
| cadena.isalpha()   | Devuelve `True` si todos los caracteres de la cadena son alfabéticos. |
| cadena.isnumeric() | Devuelve `True` si todos los caracteres de la cadena son numéricos. |
| cadena.isspace()   | Devuelve `True` si todos los caracteres de la cadena son espacios en blanco. |
| cadena.isupper()   | Devuelve `True` si todos los caracteres de la cadena están en mayúsculas. |
| cadena.islower()   | Devuelve `True` si todos los caracteres de la cadena están en minúsculas. |
| cadena.istitle()   | Devuelve `True` si la cadena de caracteres sigue las reglas de un título. |
|||
| cadena.count(cadena)          | Contar cuantas veces aparece una subcadena dada. |
| cadena.join(iterable)         | Convertir los elementos de un iterable en una cadena de caracteres. |
| cadena.replace(valor1,valor2) | Devolver una cadena de caracteres donde un valor especificado se reemplaza por otro valor. |
| cadena.split(separador)       | Dividir la cadena de caracteres en el separador especificado. |
|||
| cadena.find(cadena)                    | Buscar en la cadena un valor específico y devuelve la primera posición donde se encuentra. |
| cadena.dfind(cadena)                   | Buscar en la cadena un valor específico y devuelve la última posición donde se encuentra. |
| cadena.startswith(cadena, inicio, fin) | Devuelve `True` si la cadena de caracteres comienza en una subcadena dada. |
| cadena.endswith(cadena, inicio, fin)   | Devuelve `True` si la cadena de caracteres termina en una subcadena dada. |

</center>

```python
#Mayúsculas y Minúsculas
cadena_1 = 'hola mundo'
cadena_2 = 'HOLA MUNDO'

print('Capitalize:', cadena_1.capitalize())
print('     Upper:', cadena_1.upper())
print('     Lower:', cadena_2.lower())
print('    Titile:', cadena_1.title())
```

```python
#Verificación
cadena_1 = 'HolaMundo10'
cadena_2 = 'HOLAMUNDO'
cadena_3 = '100'
cadena_4 = 'Hola Mundo'

print('Alfanumérico:', cadena_1.isalnum())
print('  Alfabético:', cadena_2.isalpha())
print('    Numérico:', cadena_3.isnumeric())
print('     Espacio:', cadena_1.isspace())
print('   Mayúscula:', cadena_2.isupper())
print('   Minúscula:', cadena_2.islower())
print('      Título:', cadena_4.istitle())
```

```python
#Modificación
cadena_1 = 'Hola Mundo, Hola Todos'
cadena_2 = 'Hola Mundo'

print('    Contar:', cadena_1.count('Hola'))
print('      Unir:', ' '.join(['Hola','Mundo']))
print('Reemplazar:', cadena_1.replace('Todos','Amigos'))
print('   Separar:', cadena.split(' '))
```

```python
#Búsqueda
cadena = 'Hola Mundo, Hola Todos'

print('Encontrar (Izq-Der):', cadena.find('Hola'))
print('Encontrar (Der-Izq):', cadena.rfind('Hola'))
print('           Comenzar:', cadena.startswith('Hola'))
print('           Terminar:', cadena.endswith('Todos'))
```

### **2.6.3. Operaciones Básicas**

Con las cadenas de caracteres se pueden llevar a cabo dos operaciones básicas:

* Concatenación: Con el operador suma (`+`), se pueden unir dos cadenas de caracteres.

* Repetición: Con el operador multiplicación (`*`), se puede repetir una cadena de caracteres un número entero de veces.

```python
#Concateción
a = 'Hola'
b = ' Mundo'
c = a + b

print(c)
```

```python
#Repetición
r = 'Hola'*4

print(r)
```
