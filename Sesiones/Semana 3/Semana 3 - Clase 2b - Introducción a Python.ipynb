## **5.1. Introducción**

En Python, las sentencias condicionales se utilizan para ejecutar instrucciones en caso de que se cumplan una o más condiciones preestablecidas. Es una forma sencilla de permitirle a un programa tomar sus propias deciciones, ya que permite ejecutar bloques específicos de código según las necesidades presentes. Los condicionales $if$, $else$ y $elif$ acompañados de los operdores de comparación, lógicos o de identidad y pertenencia son las estructuras fundamentales para llevar a cabo el control de flujo de un programa, y su comprensión en el ámbito de la programación es muy importante, pues es finalmente lo que determina que tan dinámico y completo es éste.

### **5.2.1. If**

El conficional `if` permite que un programa ejecute instrucciones cuando se cumplan una o más condiciones. Su sintaxis es:

```
if condición:
  <<Ejecutar este bloque cuando la condición sea verdadera>>
```

Veamos:

````python
numero = 4
tipo = type(numero)

if tipo==int:
  print(f'{numero} es entero')
````

Es importante destacar que:

1. Después de la condición se agregan dos puntos (`:`).
2. El bloque de código a ejecutar si la condición se cumple debe ir con identación (tabulación) por debajo de la palabra `if`.
3. La palabra "if" en inglés traduce "si" como una condición y no "sí" como una afirmación (yes, true).

Veamos más ejemplos:

````python
numero = 4

if numero>0 and numero%2==0:
  print(f'{numero} es entero y par')
````

````python
numero = 3

if numero>0 or numero%2==0:
  print(f'{numero} es entero o par')
````

````python
frase = 'Hola Mundo'
palabra = 'Hola'

if palabra in frase:
  print(f'La palabra "{palabra}" está dentro de la frase "{frase}".')
````

Finalmente, se pueden anidar tantos `if` como sean necesarios:

````python
numero = -4
tipo = type(numero)

if tipo==int:
  print(f'{numero} es entero')

  if numero>=0:
    print(f'{numero} es natural')
  if numero<0:
    print(f'{numero} es negativo')
````

### **5.2.2. Else**

El condicional `else` permite que un programa ejecute instrucciones cuando no se cumplan una o más condiciones. Su activación depende directamente del valor de verdad asociado con la condición del `if`. Su sintaxis es:

```
if condición:
  <<Ejecutar este bloque cuando la condición sea verdadera>>
else:
  <<Ejecutar este bloque cuando la condición sea falsa>>
```

Veamos:

````python
numero = 4.5
tipo = type(numero)

if tipo==int:
  print(f'{numero} es entero')
else:
  print(f'{numero} es decimal')
````

Es importante destacar que:

1. Después del `else` se agregan dos puntos (`:`).
2. El bloque de código a ejecutar debe ir con identación (tabulación) por debajo de la palabra `else`.
3. El condicional `else` no necesita de una condición propia, pues actúa bajo la negación del conficional `if`.
4. El condicional `else` necesita del `if` para su existencia.

### **5.2.3. Elif**

El condicional `elif` permite que un programa ejecute instrucciones cuando no se cumplan una o más condiciones. Ayuda a imponer una o más condiciones ante la negación de la primera afirmación. Su activación depende directamente del valor de verdad asociado con la condición del `if`. Su sintaxis es:

```
if condición 1:
   <<Ejecutar este bloque cuando la condición 1 sea verdadera>>
elif condición 2:
   <<Ejecutar este bloque cuando la condición 2 sea verdadera>>
else:
   <<Ejecutar este bloque cuando las condiciones 1 y 2 sean falsas>
```

Veamos:

````python
numero = -4

if numero>=0:
  print(f'{numero} es positivo')
elif numero>=-10:
  print(f'{numero} es negativo mayor o igual que -10')
else:
  print(f'{numero} es negativo menor que -10')
````

O equivalentemente:

````python
numero = -4

if numero>=0:
  print(f'{numero} es positivo')

else:
  if numero>=-10:
    print(f'{numero} es negativo mayor o igual que -10')
  else:
    print(f'{numero} es negativo menor que -10')
````

 Es importante destacar que:

1. Después de la condición se agregan dos puntos (`:`).
2. El bloque de código a ejecutar debe ir con identación (tabulación) por debajo de la palabra `elif`.
3. El condicional `elif` nace de la conjunción entre `else` e `if`.
4. El condicional `elif` necesita del `if` para su existencia.

### **5.4.1. Comprimir Condicional**

Una estructura condicional tipo `if/else` puede comprimirse para ahorrar líneas de código:

````python
condicion = 1

if condicion==0: numero = 1
else: numero = 2

numero
````

````python
condicion = 1
numero = 1 if condicion==0 else 2
numero
````
