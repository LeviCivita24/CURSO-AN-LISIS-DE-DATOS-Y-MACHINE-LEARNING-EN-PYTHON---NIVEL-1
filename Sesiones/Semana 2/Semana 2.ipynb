¿Cómo se sabe en Python si un número es par?

***Ayuda:*** Puede que le sea útil el operador residuo `%`.

```python
numero = int(input('Ingrese un número: '))
respuesta = numero%2 == 0 #True or False en buleano - Tipo de dato: Bool

if respuesta==True:
  respuesta = 'Sí'

else:
  respuesta = 'No'

print(f'¿El número {numero} es par?: {respuesta}')
```

Python sabe que un número es par si n%2==0 (igual), de lo contrario será n%2!=0 (diferente).

### <font color='green'>**Ejercicio 15** </font>

Escriba un código que compruebe si un año es bisiesto o no.

***Ayuda 1:*** ¿Cuál es la regla de los años bisiesto?.   
***Ayuda 2:*** Puede que le sean útiles los operadores `or` y `and`.

```python
#Bisiesto si tiene 365+1 = 366
#2023 no lo es y 2024 sí lo es.

#Es par.
#Divisible por 4. 486484684544704, 486484684544728
```

```python
numero = 2024
div_4 = numero%4 == 0 #p: ¿Es divisble por 4? TRUE
div_100 = numero%100 == 0 #q: ¿Es divisble por 100? FALSE
div_400 = numero%400 == 0 #r: ¿Es divisble por 400? TRUE
bisiesto = div_4 and (not div_100 or div_400) #True/False

if bisiesto:
  print(f'El año {numero} es bisiesto.')
else:
  print(f'El año {numero} no es bisiesto.')
```

```python
años = [1995,1998,2000,2001,2014,2018,2020,2024]
años_bisiestos = []

for numero in años:
  div_4 = numero%4 == 0
  div_100 = numero%100 == 0
  div_400 = numero%400 == 0
  bisiesto = div_4 and (not div_100 or div_400)

  if bisiesto:
    años_bisiestos.append(numero)

años_bisiestos
```

```python
años = [1995,1998,2000,2001,2014,2018,2020,2024]
años_bisiestos = []

for numero in años:

  if numero%4 == 0 and (not numero%100 == 0 or numero%400 == 0):
    años_bisiestos.append(numero)

años_bisiestos
```
