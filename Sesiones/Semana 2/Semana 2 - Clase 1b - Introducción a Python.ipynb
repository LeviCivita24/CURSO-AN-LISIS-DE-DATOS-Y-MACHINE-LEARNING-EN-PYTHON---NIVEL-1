## **2.4. Operadores**

Los operadores hacen referencia a las expresiones que indican como se deben manipular los valores de los datos. En Python existen los siguientes tipos de operadores:

* Operadores Aritméticos
* Operadores de Comparación
* Operadores Lógicos
* Operadores de Identidad y Pertenencia
* Operadores de Asignación
* Operadores Bit a Bit (Consultar)

### **2.4.1. Operadores Aritméticos**

Estos operadores permiten ejecutar operaciones aritméticas básicas:

<center>

|Operador|Descripción|Ejemplo|
|:-:|-|:-:|
|+|  Sumar           | a + b |
|-|  Restar          | a - b |
|*|  Multiplicar     | a*b |
|/|  Dividir           | a/b |
|%|  Residuo         | a%b |
|//| División Entera | a//b |
|**| Potenciación    | a**b|

</center>

Veamos algunos ejemplo:

```python
a = 5
b = 3

print('El valor de a es:', a)
print('El valor de b es:', b)
print(f' a + b = {a+b}')
print(f' a - b = {a-b}')
print(f' a * b = {a*b}')
print(f' a / b = {a/b}')
print(f' a % b = {a%b}')
print(f'a // b = {a//b}')
print(f'a ** b = {a**b}')
```

### **2.4.2. Operadores de Comparación**

Estos operadores permiten realizar comparaciones entre cantidades. Su resultado siempre es un buleano:

<center>

|Operador|Descripción|Ejemplo|
|:-:|-|:-:|
| == | Evaluar si las cantidades son iguales. | a == b |
| != | Evaluar si las cantidades son diferentes. | a != b |
| >  | Evaluar si la cantidad de la izquierda es mayor que la de la derecha. | a > b |
| <  | Evaluar si la cantidad de la izquierda es menor que la de la derecha. | a > b |
| >= | Evaluar si la cantidad de la izquierda es mayor o igual que la de la derecha. | a >= b |
| <= | Evaluar si la cantidad de la izquierda es menor o igual que la de la derecha. | a <= b |

</center>

Veamos algunos ejemplos:

```python
a = 2
b = -2

print('El valor de a es:', a)
print('El valor de b es:', b)
print('¿a == b?:', a == b)
print('¿a != b?:', a != b)
print(' ¿a > b?:', a > b)
print(' ¿a < b?:', a < b)
print('¿a >= b?:', a >= b)
print('¿a <= b?:', a <= b)
```

### **2.4.3. Operadores Lógicos**

Estos operadores permiten comparar afirmaciones. Su resultado siempre es un buleano:

<center>

|Operador|Descripción|Ejemplo|
|:-:|-|:-:|
| and | Devuelve verdadero si las dos afirmaciones son verdaderas. | a and b |
| or  | Devuelve falso si las dos afirmaciones son falsas.         | a or b |
| not | Negar la afimación.                                        | not a |

</center>

Veamos algunos ejemplos:

```python
True and False
```

```python
True or False
```

```python
not False
```

```python
cantidad = 2.0
a = cantidad%1 == 0
b = 0 <= cantidad < 5
c = cantidad%2 != 0

print('¿Es un entero?', a)
print('¿Está entre [0,5)?', b)
print('¿No es divisible por 2?', c)
print('¿Es un digito?', a and b)
print('¿No es par?', not a or not b)
```

### **2.4.4. Operadores de Identidad y Pertenencia**

Estos operadores permiten comparar entre las direcciones y los contenidos de las cantidades. Su resultado siempre es un buleano:

<center>

|Operador|Descripción|Ejemplo|
|:-:|-|:-:|
| is     | Evaluar si las direcciones de las cantidades son iguales (si son el mismo objeto). | a is b |
| is not | Evaluar si las direcciones de las cantidades no son iguales (si son el diferentes objetos). | a is not b |
| in     | Evaluar si un valor pertenece a otro. | a in b |
| not in | Evaluar si un valor no pertenece a otro. | a not in b |

</center>

Veamos algunos ejemplos:

```python
a = [1,2]
b = [1,2]
c = 1

print('        ¿a es b?:', a is b)
print('     ¿a no es b?:', a is not b)
print('   ¿c está en a?:', c in a)
print('¿c no está en a?:', c not in a)
```

```python
a = ['Yeison','Dayanna','Alejandro']
c = ['Yeison','Dayanna','Alejandro']

print(c is a)
print(c in a)
print('Yeison' in a)
print(c == a)
```

### **2.4.5. Operadores de Asignación**

Estos operadores permiten asignar valores a variables:

<center>

|Operador|Descripción|Ejemplo|
|:-:|-|:-:|
| =   | Asignar el valor de la derecha a la variable de la izquierda.      | a = 2 |
| +=  | Sumar el valor de la derecha a la variable de la izquierda.        | a += 2 |
| -=  | Restar el valor de la derecha a la variable de la izquierda.       | a -= 2 |
| *=  | Multiplicar el valor de la derecha a la variable de la izquierda.  | a *= 2 |
| /=  | Dividir la variable de la izquierda por el valor de la derecha.    | a /= 2 |
| %=  | Calcular el residuo entre la variable de la izquierda y el valor de la derecha. | a %= 2 |
| //= | Dividir en forma entera la variable de la izquierda por el valor de la derecha. | a //= 2 |
| **= | Potenciar la variable de la izquierda por el valor de la derecha.  | a **= 2 |

</center>

Veamos algunos ejemplos:

```python
a = 2
print(a)

a = a+2
print(a)
```

```python
a = 2
print(a)

a += 2
print(a)
```

Nótese que los dos códigos anteriores efectúan en lismo procedimiento.
